# Q. Analytical Derivation

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$

Answer the followings:

1. Show that
$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-),$$
where
$$m_k^- = A_{k-1}m_{k-1},$$
and
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

2. Show that
$$y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m).$$

3. Show that
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

4. Show that
$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$
where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

5. Find
$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$  and
$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$.

To derive the joint distribution of two Gaussian random variables, we can form a new vector $Z = \begin{bmatrix} x_k^- \ y_k^- \end{bmatrix}$$Z = \begin{bmatrix} x_k^- \ y_k^- \end{bmatrix}$.

$E[Z] = E\begin{bmatrix} x_k^- \ y_k^- \end{bmatrix} = \begin{bmatrix} E[x_k^-] \ E[y_k^-] \end{bmatrix} = \begin{bmatrix} m_k^- \ H_k m_k^- \end{bmatrix}$$E[Z] = E\begin{bmatrix} x_k^- \ y_k^- \end{bmatrix} = \begin{bmatrix} E[x_k^-] \ E[y_k^-] \end{bmatrix} = \begin{bmatrix} m_k^- \ H_k m_k^- \end{bmatrix}$.

# Q. 1-D Example

Consider the scalar linear-Gaussian filter model
\begin{aligned}
x^-_k &= a\,x^+_{k-1} + w_{k-1},\qquad w_{k-1}\sim\mathscr N(0,q),\\
y^-_k &= h\,x^-_k + z_k,\qquad\;\;\;\; z_k\sim\mathscr N(0,r),
\end{aligned}
with prior $x_0\sim \mathscr{N}(m_0,P_0)$. Define $Y_{k}=\{y^{\mathrm{obs}}_1,\dots,y^{\mathrm{obs}}_{k}\}$.

Answer the following:

1. Show that
\begin{aligned}
m_k^- &= a\,m_{k-1},\\
P_k^- &= a^2 P_{k-1} + q.
\end{aligned}

2. Show that
\begin{aligned}
m_k &= m_k^- + K_k\,v_k
= m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h\,m_k^-\bigr),\\
P_k &= (1 - K_k h)\,P_k^-
= \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^- .
\end{aligned}

3. Show that the predictive measurement distribution (before seeing $y_k$):
\begin{align*}
p(y^-_k | Y_{k-1})=\mathscr N\bigl(h\,m_k^-,\; h^2 P_k^- + r\bigr).
\end{align*}

4. Show that the posterior-predictive measurement distribution (after filtering on $y_k$):
\begin{align*}
p(y^-_k\mid Y_k)=\mathscr N\bigl(h\,m_k,\; h^2 P_k + r\bigr).
\end{align*}

5. For suitable numerical values for $a,q,h,r$ animate the above to prior distribution and the posterior distribution.

1.Given the state transition equation: $x_k^- = a x_{k-1}^+ + w_{k-1}$$x_k^- = a x_{k-1}^+ + w_{k-1}$, where $x_{k-1}^+ \sim \mathscr{N}(m_{k-1}, P_{k-1})$$x_{k-1}^+ \sim \mathscr{N}(m_{k-1}, P_{k-1})$ and $w_{k-1} \sim \mathscr{N}(0, q)$$w_{k-1} \sim \mathscr{N}(0, q)$. We assume $x_{k-1}^+$$x_{k-1}^+$ and $w_{k-1}$$w_{k-1}$ are independent.

Mean ($m_k^-$$m_k^-$): The expected value of $x_k^-$ is: $E[x_k^-] = E[a x_{k-1}^+ + w_{k-1}]$ By linearity of expectation: $E[x_k^-] = a E[x_{k-1}^+] + E[w_{k-1}]$ Substitute the given means: $m_k^- = a m_{k-1} + 0 = a m_{k-1}$.

Covariance ($P_k^-$$P_k^-$): The variance of $x_k^-$ is: $\text{Var}(x_k^-) = \text{Var}(a x_{k-1}^+ + w_{k-1})$ Since $x_{k-1}^+$ and $w_{k-1}$ are independent, the variance of their sum is the sum of their variances: $\text{Var}(x_k^-) = \text{Var}(a x_{k-1}^+) + \text{Var}(w_{k-1})$ For a constant 'a', $\text{Var}(aX) = a^2 \text{Var}(X)$: $\text{Var}(x_k^-) = a^2 \text{Var}(x_{k-1}^+) + \text{Var}(w_{k-1})$ Substitute the given variances: $P_k^- = a^2 P_{k-1} + q$.

Thus, the predicted state $x_k^-$$x_k^-$ follows a Gaussian distribution $\mathscr{N}(a m_{k-1}, a^2 P_{k-1} + q)$$\mathscr{N}(a m_{k-1}, a^2 P_{k-1} + q)$.

2.Given the measurement equation: $y_k^- = h x_k^- + z_k$$y_k^- = h x_k^- + z_k$, where $z_k \sim \mathscr{N}(0, r)$$z_k \sim \mathscr{N}(0, r)$. We've already established $x_k^- \sim \mathscr{N}(m_k^-, P_k^-)$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-)$.

From the general Kalman filter update equations:

Kalman Gain ($K_k$$K_k$): In the scalar case, $H_k = h$, $\Sigma_m = r$. The innovation covariance $S_k = H_k P_k^- H_k^T + \Sigma_m$ becomes $S_k = h P_k^- h + r = h^2 P_k^- + r$. The Kalman gain $K_k = P_k^- H_k^T (S_k)^{-1}$ becomes: $K_k = P_k^- h (h^2 P_k^- + r)^{-1} = \frac{P_k^- h}{h^2 P_k^- + r} = \frac{P_k^- h}{S_k}$.

Updated Mean ($m_k$$m_k$): The updated mean is $m_k = m_k^- + K_k(y^{\mathrm{obs}}_k - H_k m_k^-)$. Substitute $H_k=h$ and the expression for $K_k$: $m_k = m_k^- + K_k (y^{\mathrm{obs}}_k - h m_k^-)$. Defining the innovation $v_k = (y^{\mathrm{obs}}_k - h m_k^-)$, we get $m_k = m_k^- + K_k v_k$. Substituting $K_k = \frac{P_k^- h}{S_k}$: $m_k = m_k^- + \frac{P_k^- h}{S_k}(y^{\mathrm{obs}}_k - h m_k^-)$.

Updated Covariance ($P_k$$P_k$): The updated covariance is $P_k = (I - K_k H_k)P_k^-$. For the scalar case, $I=1$ and $H_k=h$: $P_k = (1 - K_k h)P_k^-$. Substituting $K_k = \frac{P_k^- h}{S_k}$: $P_k = (1 - \frac{P_k^- h}{S_k} h)P_k^- = (1 - \frac{P_k^- h^2}{S_k})P_k^-$.

3.This is the distribution of the measurement $y_k^-$$y_k^-$ based on the predicted state $x_k^-$$x_k^-$ before the actual measurement $y_k^{\mathrm{obs}}$$y_k^{\mathrm{obs}}$ is incorporated. We use the measurement model $y_k^- = h x_k^- + z_k$$y_k^- = h x_k^- + z_k$.

Mean: $E[y_k^- | Y_{k-1}] = E[h x_k^- + z_k | Y_{k-1}]$ Since $x_k^- \sim \mathscr{N}(m_k^-, P_k^-)$ and $z_k \sim \mathscr{N}(0, r)$ are independent: $E[y_k^- | Y_{k-1}] = h E[x_k^-] + E[z_k] = h m_k^- + 0 = h m_k^-$.

Variance: $\text{Var}(y_k^- | Y_{k-1}) = \text{Var}(h x_k^- + z_k | Y_{k-1})$ $= \text{Var}(h x_k^-) + \text{Var}(z_k)$ $= h^2 \text{Var}(x_k^-) + r = h^2 P_k^- + r$.

Since $x_k^-$$x_k^-$ and $z_k$$z_k$ are Gaussian, their linear combination $y_k^-$$y_k^-$ is also Gaussian. Thus, $p(y_k^- | Y_{k-1}) = \mathscr{N}(h m_k^-, h^2 P_k^- + r)$$p(y_k^- | Y_{k-1}) = \mathscr{N}(h m_k^-, h^2 P_k^- + r)$.

4.This distribution refers to what we expect a measurement $y_k^-$$y_k^-$ to be, given the updated state estimate $x_k^+ \sim \mathscr{N}(m_k, P_k)$$x_k^+ \sim \mathscr{N}(m_k, P_k)$ after incorporating $y_k^{\mathrm{obs}}$$y_k^{\mathrm{obs}}$. Essentially, it's the distribution of $y_k^-$$y_k^-$ if we were to take another measurement at time $k$$k$, but now with the benefit of the refined state estimate.

Mean: $E[y_k^- | Y_k] = E[h x_k^+ + z_k | Y_k]$ Using the updated state $x_k^+ \sim \mathscr{N}(m_k, P_k)$ and $z_k \sim \mathscr{N}(0, r)$ (independent): $E[y_k^- | Y_k] = h E[x_k^+] + E[z_k] = h m_k + 0 = h m_k$.

Variance: $\text{Var}(y_k^- | Y_k) = \text{Var}(h x_k^+ + z_k | Y_k)$ $= \text{Var}(h x_k^+) + \text{Var}(z_k)$ $= h^2 \text{Var}(x_k^+) + r = h^2 P_k + r$.

Since $x_k^+$$x_k^+$ and $z_k$$z_k$ are Gaussian, their linear combination $y_k^-$$y_k^-$ is also Gaussian. Thus, $p(y_k^- | Y_k) = \mathscr{N}(h m_k, h^2 P_k + r)$$p(y_k^- | Y_k) = \mathscr{N}(h m_k, h^2 P_k + r)$.

3.  **Measurement Likelihood:** $\mathscr{N}(y^{\mathrm{obs}}_k, r)$ (the distribution representing the noisy measurement)
4.  **Posterior Distribution:** $\mathscr{N}(m_k, P_k)$ (the state after measurement update)


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy.stats import norm
from IPython.display import HTML

# --- 1D Kalman Filter Parameters ---
# State transition constant 'a'
a = 0.95 # Slightly decaying system
# Process noise variance 'q'
q = 0.1**2 # Small process noise
# Measurement constant 'h'
h = 1.0 # Direct measurement of the state
# Measurement noise variance 'r'
r = 0.5**2 # Moderate measurement noise

# Number of simulation steps
num_steps = 50

# Initial state parameters
m0 = 0.0  # Initial mean
P0 = 1.0  # Initial variance (high uncertainty)

# --- Simulate True State and Measurements ---
true_states = np.zeros(num_steps)
measurements = np.zeros(num_steps)

current_true_state = np.random.normal(m0, np.sqrt(P0)) # Initial true state

for k in range(num_steps):
    # True state evolution
    process_noise = np.random.normal(0, np.sqrt(q))
    current_true_state = a * current_true_state + process_noise
    true_states[k] = current_true_state

    # Generate noisy measurement
    measurement_noise = np.random.normal(0, np.sqrt(r))
    measurements[k] = h * current_true_state + measurement_noise

print("Simulation data generated.")
```

```python
# --- Kalman Filter Implementation ---
state_means = np.zeros(num_steps)
state_covariances = np.zeros(num_steps)
predicted_means = np.zeros(num_steps)
predicted_covariances = np.zeros(num_steps)

# Initial values for Kalman Filter
m_prev = m0
P_prev = P0

for k in range(num_steps):
    # Prediction Step
    m_minus = a * m_prev
    P_minus = a**2 * P_prev + q

    predicted_means[k] = m_minus
    predicted_covariances[k] = P_minus

    # Update Step
    # Innovation covariance (S_k)
    S_k = h**2 * P_minus + r
    # Kalman Gain (K_k)
    K_k = (P_minus * h) / S_k

    # Innovation (v_k)
    v_k = measurements[k] - h * m_minus

    # Updated mean (m_k)
    m_current = m_minus + K_k * v_k
    # Updated covariance (P_k)
    P_current = (1 - K_k * h) * P_minus

    state_means[k] = m_current
    state_covariances[k] = P_current

    # Set current to previous for next iteration
    m_prev = m_current
    P_prev = P_current

print("Kalman filter applied.")
```

```python
# --- Animation Setup ---
fig, ax = plt.subplots(figsize=(12, 7))
ax.grid(True)
ax.set_xlabel('State Value (x)')
ax.set_ylabel('Probability Density')
ax.set_title('1D Kalman Filter: Prior, Predicted, Likelihood, Posterior')

# Set x-limits for the plot based on true states and measurements
min_val = min(np.min(true_states), np.min(measurements))
max_val = max(np.max(true_states), np.max(measurements))
x_range = np.linspace(min_val - 2, max_val + 2, 500)

ax.set_xlim(x_range[0], x_range[-1])
ax.set_ylim(0, 2.5) # Adjust y-limit as needed

# Lines for distributions
line_prior, = ax.plot([], [], ':', color='orange', label='Prior (State k-1)', alpha=0.7)
line_predicted, = ax.plot([], [], '--', color='purple', label='Predicted State (k)', alpha=0.7)
line_likelihood, = ax.plot([], [], '-', color='gray', label='Measurement Likelihood (k)', alpha=0.7)
line_posterior, = ax.plot([], [], '-', color='blue', label='Posterior (State k)', alpha=0.9, linewidth=2)

# Markers for true state, measurement, and estimate
marker_true, = ax.plot([], [], 'o', color='green', markersize=8, label='True State')
marker_measurement, = ax.plot([], [], 'x', color='red', markersize=10, label='Measurement')
marker_estimate, = ax.plot([], [], '^', color='blue', markersize=10, label='KF Estimate')

ax.legend()

def animate(i):
    # Update Prior
    if i == 0:
        prior_mean = m0
        prior_cov = P0
    else:
        prior_mean = state_means[i-1]
        prior_cov = state_covariances[i-1]
    line_prior.set_data(x_range, norm.pdf(x_range, prior_mean, np.sqrt(prior_cov)))

    # Update Predicted State Distribution
    predicted_pdf = norm.pdf(x_range, predicted_means[i], np.sqrt(predicted_covariances[i]))
    line_predicted.set_data(x_range, predicted_pdf)

    # Update Measurement Likelihood
    likelihood_pdf = norm.pdf(x_range, measurements[i], np.sqrt(r)) # N(y_obs, r)
    line_likelihood.set_data(x_range, likelihood_pdf)

    # Update Posterior
    posterior_pdf = norm.pdf(x_range, state_means[i], np.sqrt(state_covariances[i]))
    line_posterior.set_data(x_range, posterior_pdf)

    # Update Markers
    marker_true.set_data([true_states[i]], [0])
    marker_measurement.set_data([measurements[i]], [0])
    marker_estimate.set_data([state_means[i]], [0])

    ax.set_title(f'1D Kalman Filter (Step {i+1}/{num_steps}): Prior, Predicted, Likelihood, Posterior')

    return line_prior, line_predicted, line_likelihood, line_posterior, marker_true, marker_measurement, marker_estimate

ani = animation.FuncAnimation(fig, animate, frames=num_steps, interval=200, blit=True)

# To display in Colab, you might need to convert to HTML
HTML(ani.to_jshtml())

# If you want to save it as a GIF (requires imagemagick or ffmpeg)
# ani.save('kalman_1d.gif', writer='imagemagick', fps=5)
print("Animation generated. If it doesn't display directly, run the cell to see the HTML output.")
```

## Part B

Develop a python based method to filter a sequence of noisy GPS obtained position meaurements.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time step
dt = 1.0

# State transition matrix A
A = np.array([
    [1, 0, dt, 0],
    [0, 1, 0, dt],
    [0, 0, 1, 0],
    [0, 0, 0, 1]
])

# Measurement matrix H
H = np.array([
    [1, 0, 0, 0],
    [0, 1, 0, 0]
])

# Process noise input matrix G
G = np.array([
    [0.5 * dt**2, 0],
    [0, 0.5 * dt**2],
    [dt, 0],
    [0, dt]
])

# Process noise covariance Sigma_p (for w_k-1)
# Assuming independent acceleration noise components in x and y
process_noise_std = 0.1 # Standard deviation of acceleration noise
Sigma_p = np.eye(2) * process_noise_std**2 # 2x2 covariance for w_k-1

# Process noise covariance Q (for state x_k)
Q = G @ Sigma_p @ G.T

# Measurement noise covariance Sigma_m (for z_k)
# Assuming independent position measurement noise components in x and y
measurement_noise_std = 1.0 # Standard deviation of position measurement noise
Sigma_m = np.eye(2) * measurement_noise_std**2 # 2x2 covariance for z_k

print("A matrix:\n", A)
print("\nH matrix:\n", H)
print("\nG matrix:\n", G)
print("\nProcess noise covariance Q:\n", Q)
print("\nMeasurement noise covariance Sigma_m:\n", Sigma_m)
```

```python
def simulate_data(num_steps, initial_state, A, G, Sigma_p, H, Sigma_m):
    state_dim = initial_state.shape[0]
    measurement_dim = H.shape[0]

    true_states = np.zeros((num_steps, state_dim))
    measurements = np.zeros((num_steps, measurement_dim))

    current_state = initial_state

    for k in range(num_steps):
        # Simulate process noise
        w_k_minus_1 = np.random.multivariate_normal(np.zeros(Sigma_p.shape[0]), Sigma_p)

        # True state evolution
        current_state = A @ current_state + G @ w_k_minus_1
        true_states[k] = current_state

        # Simulate measurement noise
        z_k = np.random.multivariate_normal(np.zeros(Sigma_m.shape[0]), Sigma_m)

        # Generate noisy measurement
        measurements[k] = H @ current_state + z_k

    return true_states, measurements

# Initial true state: [px, py, vx, vy]
initial_true_state = np.array([0.0, 0.0, 1.0, 0.5]) # Start at (0,0) with some velocity
num_simulation_steps = 100

true_states, noisy_measurements = simulate_data(
    num_simulation_steps, initial_true_state, A, G, Sigma_p, H, Sigma_m
)

print(f"Generated {num_simulation_steps} true states and noisy measurements.")
# display(true_states[:5, :])
# display(noisy_measurements[:5, :])
```

```python
def kalman_filter(A, Q, H, R, initial_state_estimate, initial_covariance, measurements):
    state_dim = A.shape[0]
    num_steps = measurements.shape[0]

    # Initialize lists to store estimates
    state_estimates = np.zeros((num_steps, state_dim))
    covariance_estimates = np.zeros((num_steps, state_dim, state_dim))

    # Initial belief
    x_hat = initial_state_estimate
    P = initial_covariance

    for k in range(num_steps):
        # --- Prediction Step ---
        # Predicted state estimate (m_k_minus)
        x_hat_minus = A @ x_hat

        # Predicted covariance estimate (P_k_minus)
        P_minus = A @ P @ A.T + Q

        # --- Update Step ---
        # Kalman Gain (K_k)
        S = H @ P_minus @ H.T + R # Innovation (or residual) covariance
        K = P_minus @ H.T @ np.linalg.inv(S)

        # Updated state estimate (m_k)
        y_observed = measurements[k]
        innovation = y_observed - H @ x_hat_minus
        x_hat = x_hat_minus + K @ innovation

        # Updated covariance estimate (P_k)
        P = (np.eye(state_dim) - K @ H) @ P_minus

        state_estimates[k] = x_hat
        covariance_estimates[k] = P

    return state_estimates, covariance_estimates

# Initial Kalman Filter parameters
initial_kf_state_estimate = np.array([0.0, 0.0, 0.0, 0.0]) # Initial guess for state [px, py, vx, vy]
# Initial covariance: High uncertainty in position and velocity
initial_kf_covariance = np.eye(4) * 100.0

# Run the Kalman filter
filtered_states, filtered_covariances = kalman_filter(
    A, Q, H, Sigma_m, initial_kf_state_estimate, initial_kf_covariance, noisy_measurements
)

print(f"Kalman filter ran for {num_simulation_steps} steps.")
# display(filtered_states[:5, :])
```
