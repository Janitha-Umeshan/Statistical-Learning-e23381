### 1. The Partition Function

The canonical partition function is given by:

$$Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$$

Substitute the energy eigenvalues $E_j = \hbar \omega \left(j + \frac{1}{2}\right)$:

$$Z(T) = \sum_{j=0}^\infty e^{-\frac{\hbar \omega (j + 1/2)}{k_B T}}$$

$$Z(T) = \sum_{j=0}^\infty e^{-\frac{\hbar \omega j}{k_B T}} e^{-\frac{\hbar \omega}{2k_B T}}$$

Factor out the term that does not depend on $j$:

$$Z(T) = e^{-\frac{\hbar \omega}{2k_B T}} \sum_{j=0}^\infty \left(e^{-\frac{\hbar \omega}{k_B T}}\right)^j$$

Let $x = e^{-\frac{\hbar \omega}{k_B T}}$. Since $\hbar, \omega, k_B, T > 0$, we have $0 < x < 1$. The sum is a geometric series:

$$\sum_{j=0}^\infty x^j = \frac{1}{1 - x}$$

Substituting $x$ back:

$$Z(T) = e^{-\frac{\hbar \omega}{2k_B T}} \frac{1}{1 - e^{-\frac{\hbar \omega}{k_B T}}}$$

Thus, the partition function is:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}}$$

### 2. Thermodynamic Potentials

$$\ln Z = \ln\left(\frac{e^{-\frac{\hbar \omega}{2k_B T}}}{1 - e^{-\frac{\hbar \omega}{k_B T}}}\right)$$

$$\ln Z = -\frac{\hbar \omega}{2k_B T} - \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$$

#### Helmholtz Free Energy, $F(T)$

$$F(T) = -k_B T \ln Z$$

$$F(T) = -k_B T \left(-\frac{\hbar \omega}{2k_B T} - \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)\right)$$

$$F(T) = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$$

#### Internal Energy, $U(T)$

The internal energy is given by $U(T) = -\frac{\partial \ln Z}{\partial \beta}$, where $\beta = \frac{1}{k_B T}$.

Let's rewrite $\ln Z$ in terms of $\beta$:

$$\ln Z = -\frac{\hbar \omega \beta}{2} - \ln(1 - e^{-\hbar \omega \beta})$$

Now, differentiate with respect to $\beta$:

$$\frac{\partial \ln Z}{\partial \beta} = -\frac{\hbar \omega}{2} - \frac{1}{1 - e^{-\hbar \omega \beta}} \cdot (-\hbar \omega) e^{-\hbar \omega \beta}$$

$$\frac{\partial \ln Z}{\partial \beta} = -\frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\hbar \omega \beta}}{1 - e^{-\hbar \omega \beta}}$$

$$U(T) = -\left(-\frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\hbar \omega \beta}}{1 - e^{-\hbar \omega \beta}}\right)$$

### 3. Entropy, $S(T)$

The entropy is given by $S = -\left(\frac{\partial F}{\partial T}\right)$.

We have $F(T) = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$.

Let $x = \frac{\hbar \omega}{k_B T}$. Then $F(T) = \frac{\hbar \omega}{2} + k_B T \ln(1 - e^{-x})$.

Now, differentiate $F(T)$ with respect to $T$:

$$\frac{\partial F}{\partial T} = \frac{\partial}{\partial T}\left(\frac{\hbar \omega}{2}\right) + \frac{\partial}{\partial T}\left(k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)\right)$$

The first term is 0. For the second term, use the product rule:

$$\frac{\partial F}{\partial T} = k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + k_B T \frac{\partial}{\partial T}\left(\ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)\right)$$

To differentiate $\ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$ with respect to $T$:

Let $u = 1 - e^{-\frac{\hbar \omega}{k_B T}}$. Then $\frac{d}{dT} \ln(u) = \frac{1}{u} \frac{du}{dT}$.

$$\frac{du}{dT} = - e^{-\frac{\hbar \omega}{k_B T}} \cdot \frac{\partial}{\partial T}\left(-\frac{\hbar \omega}{k_B T}\right)$$

$$\frac{\partial}{\partial T}\left(-\frac{\hbar \omega}{k_B T}\right) = -\frac{\hbar \omega}{k_B} \cdot (-T^{-2}) = \frac{\hbar \omega}{k_B T^2}$$

So, $\frac{du}{dT} = - e^{-\frac{\hbar \omega}{k_B T}} \frac{\hbar \omega}{k_B T^2}$.

Therefore,

$$\frac{\partial}{\partial T}\left(\ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)\right) = \frac{1}{1 - e^{-\frac{\hbar \omega}{k_B T}}} \cdot \left(- e^{-\frac{\hbar \omega}{k_B T}} \frac{\hbar \omega}{k_B T^2}\right)$$

Substitute this back into the derivative of $F(T)$:

$$\frac{\partial F}{\partial T} = k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + k_B T \left(\frac{-\hbar \omega e^{-\frac{\hbar \omega}{k_B T}}}{k_B T^2 (1 - e^{-\frac{\hbar \omega}{k_B T}})}\right)$$

$$\frac{\partial F}{\partial T} = k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) - \frac{\hbar \omega e^{-\frac{\hbar \omega}{k_B T}}}{T (1 - e^{-\frac{\hbar \omega}{k_B T}})}$$

Now, $S(T) = -\left(\frac{\partial F}{\partial T}\right)$:

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega e^{-\frac{\hbar \omega}{k_B T}}}{T (1 - e^{-\frac{\hbar \omega}{k_B T}})}$$

We can rewrite the second term by dividing numerator and denominator by $e^{-\frac{\hbar \omega}{k_B T}}$:

$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{T (e^{\frac{\hbar \omega}{k_B T}} - 1)}$$

#### Verification: $U = F + TS$

We have:

$$F = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)$$

$$TS = -k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + T \frac{\hbar \omega}{T (e^{\frac{\hbar \omega}{k_B T}} - 1)}$$

$$TS = -k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}$$

Now, sum $F + TS$:

$$F + TS = \left(\frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right)\right) + \left(-k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_B T}}\right) + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}\right)$$

The $k_B T \ln$ terms cancel out:

$$F + TS = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}$$

This is indeed equal to our expression for $U(T)$. So, $U = F + TS$ is verified.

### 4. Heat Capacity and Hyperbolic Form

#### Calculate the constant-volume Heat Capacity, $C_V = \left(\frac{\partial U}{\partial T}\right)$

We have $U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}$.

Let $x = \frac{\hbar \omega}{k_B T}$. Then $U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^x - 1}$.

Differentiate with respect to $T$:

$$C_V = \frac{\partial}{\partial T}\left(\frac{\hbar \omega}{2}\right) + \frac{\partial}{\partial T}\left(\frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}\right)$$

The first term is 0.

Let $f(T) = \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_B T}} - 1}$. Use the chain rule.

Let $u = e^{\frac{\hbar \omega}{k_B T}} - 1$. Then $\frac{d}{dT} (\hbar \omega u^{-1}) = -\hbar \omega u^{-2} \frac{du}{dT}$.

First, find $\frac{du}{dT}$:

$$\frac{du}{dT} = \frac{\partial}{\partial T}\left(e^{\frac{\hbar \omega}{k_B T}} - 1\right) = e^{\frac{\hbar \omega}{k_B T}} \cdot \frac{\partial}{\partial T}\left(\frac{\hbar \omega}{k_B T}\right)$$

$$\frac{\partial}{\partial T}\left(\frac{\hbar \omega}{k_B T}\right) = -\frac{\hbar \omega}{k_B T^2}$$

So, $\frac{du}{dT} = -e^{\frac{\hbar \omega}{k_B T}} \frac{\hbar \omega}{k_B T^2}$.

Now, substitute this back:

$$C_V = -\hbar \omega \left(e^{\frac{\hbar \omega}{k_B T}} - 1\right)^{-2} \left(-e^{\frac{\hbar \omega}{k_B T}} \frac{\hbar \omega}{k_B T^2}\right)$$

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{e^{\frac{\hbar \omega}{k_B T}}}{(e^{\frac{\hbar \omega}{k_B T}} - 1)^2}$$

#### Hyperbolic Identity

We need to prove that $C_V$ can be rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_B T}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_B T}\right)}$$

Let $y = \frac{\hbar \omega}{2k_B T}$. Then $\frac{\hbar \omega}{k_B T} = 2y$.

From the identity $\sinh(y) = \frac{e^y - e^{-y}}{2}$, we have $\sinh^2(y) = \frac{(e^y - e^{-y})^2}{4}$.

Also, we know that $e^{2y} - 1 = (e^y - e^{-y}) e^y = 2e^y \sinh(y)$.

Let's manipulate our expression for $C_V$:

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{e^{\frac{\hbar \omega}{k_B T}}}{(e^{\frac{\hbar \omega}{k_B T}} - 1)^2}$$

Substitute $2y = \frac{\hbar \omega}{k_B T}$:

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{e^{2y}}{(e^{2y} - 1)^2}$$

We know that $e^{2y} - 1 = (e^y - e^{-y})(e^y + e^{-y}) = 2 \sinh(y) \cdot 2 \cosh(y) = 4 \sinh(y) \cosh(y)$.

Also, $e^{2y} - 1 = (e^y - e^{-y})e^y + e^y - e^y = e^y (e^y - e^{-y})$.

Let's use a simpler manipulation: divide the numerator and denominator of the fraction $\frac{e^{2y}}{(e^{2y} - 1)^2}$ by $e^{2y}$.

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{1}{(e^y - e^{-y})^2}$$

We know that $e^y - e^{-y} = 2 \sinh(y)$. So $(e^y - e^{-y})^2 = 4 \sinh^2(y)$.

$$C_V = \frac{(\hbar \omega)^2}{k_B T^2} \frac{1}{4 \sinh^2(y)}$$

Rearrange the terms:

$$C_V = k_B \frac{(\hbar \omega)^2}{4 k_B^2 T^2} \frac{1}{\sinh^2(y)}$$

$$C_V = k_B \left(\frac{\hbar \omega}{2k_B T}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_B T}\right)}$$

This matches the desired hyperbolic form.

### 5. Asymptotic Behavior (Discussion)

We are interested in the high-temperature limit, where $k_B T \gg \hbar \omega$. This means that the argument of the hyperbolic sine function, $y = \frac{\hbar \omega}{2k_B T}$, becomes very small ($y \ll 1$).

For small $y$, we can use the approximation $\sinh(y) \approx y$.

Substitute this into the expression for $C_V$:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_B T}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_B T}\right)}$$

$$C_V \approx k_B \left(\frac{\hbar \omega}{2k_B T}\right)^2 \frac{1}{\left(\frac{\hbar \omega}{2k_B T}\right)^2}$$

$$C_V \approx k_B$$

#### Classical Equipartition Theorem Prediction

For a classical harmonic oscillator, there are two degrees of freedom associated with energy: one for kinetic energy (proportional to $p^2$) and one for potential energy (proportional to $q^2$). According to the equipartition theorem, each quadratic degree of freedom contributes $\frac{1}{2}k_B T$ to the internal energy. Therefore, a classical one-dimensional harmonic oscillator has an internal energy of $U_{classical} = k_B T$.

The classical heat capacity at constant volume would then be:

$$C_{V, classical} = \left(\frac{\partial U_{classical}}{\partial T}\right) = \frac{\partial (k_B T)}{\partial T} = k_B$$

#### Comparison

Yes, the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$) matches the classical equipartition theorem prediction. At high temperatures, the energy levels of the quantum oscillator are so closely spaced relative to the thermal energy $k_B T$ that the system behaves classically, and each degree of freedom contributes $k_B T$ to the internal energy, leading to a heat capacity of $k_B$ per oscillator.

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

The problem asks to find the probability density function $p(x)$ for a continuous random variable $X \in (-\infty, \infty)$ that maximizes entropy, subject to given constraints on its mean and variance.

### 1. Define Entropy
For a continuous random variable, the differential entropy $H(p)$ is given by:
$$H(p) = -\int_{-\infty}^{\infty} p(x) \ln p(x) dx$$
Our goal is to maximize this quantity.

### 2. Define Constraints
We have three constraints:

*   **Normalization Constraint:** The total probability must be 1.
    $$\int_{-\infty}^{\infty} p(x) dx = 1$$

*   **Mean Constraint:** The expected value of $X$ is $\mu$.
    $$\mathbb{E}[X] = \int_{-\infty}^{\infty} x p(x) dx = \mu$$

*   **Variance Constraint:** The expected value of $(X-\mu)^2$ is $\sigma^2$.
    $$\mathbb{E}[(X-\mu)^2] = \int_{-\infty}^{\infty} (x-\mu)^2 p(x) dx = \sigma^2$$


### 3. Set up the Lagrangian
To maximize the entropy subject to these constraints, we use the method of Lagrange multipliers. The Lagrangian $L$ is defined as:

$$L = -\int_{-\infty}^{\infty} p(x) \ln p(x) dx - \lambda_0 \left(\int_{-\infty}^{\infty} p(x) dx - 1\right) - \lambda_1 \left(\int_{-\infty}^{\infty} x p(x) dx - \mu\right) - \lambda_2 \left(\int_{-\infty}^{\infty} (x-\mu)^2 p(x) dx - \sigma^2\right)$$

where $\lambda_0$, $\lambda_1$, and $\lambda_2$ are the Lagrange multipliers corresponding to the normalization, mean, and variance constraints, respectively.

### 4. Take the Variational Derivative
To find the function $p(x)$ that maximizes $L$, we take the variational derivative with respect to $p(x)$ and set it to zero:

$$\frac{\delta L}{\delta p(x)} = -\left(\ln p(x) + 1\right) - \lambda_0 - \lambda_1 x - \lambda_2 (x-\mu)^2 = 0$$

Rearranging the terms to solve for $\ln p(x)$:

$$\ln p(x) = -1 - \lambda_0 - \lambda_1 x - \lambda_2 (x-\mu)^2$$

Exponentiating both sides, we get the form of $p(x)$:

$$p(x) = e^{-1 - \lambda_0 - \lambda_1 x - \lambda_2 (x-\mu)^2}$$

Let $A = e^{-1 - \lambda_0}$. Then,

$$p(x) = A \exp\left( -\lambda_1 x - \lambda_2 (x-\mu)^2 \right)$$

This expression gives the general form of the probability density function that maximizes entropy under the given constraints. Now, we need to determine the specific values of the Lagrange multipliers $A$, $\lambda_1$, and $\lambda_2$ using the constraint equations.

### 5. Solve for the Lagrange Multipliers

Let's expand the exponent:

$$p(x) = A \exp\left( -\lambda_1 x - \lambda_2 (x^2 - 2\mu x + \mu^2) \right)$$

$$p(x) = A \exp\left( -\lambda_2 x^2 + (2\mu\lambda_2 - \lambda_1) x - \lambda_2 \mu^2 \right)$$

To simplify, let's complete the square in the exponent. The general form of a Gaussian distribution is $C \exp\left( -\frac{(x - \nu)^2}{2\tau^2} \right)$.

Matching the $x^2$ term, we must have:

$$-\lambda_2 x^2 = -\frac{1}{2\tau^2} x^2 \implies \lambda_2 = \frac{1}{2\tau^2}$$

Matching the $x$ term, we must have:

$$(2\mu\lambda_2 - \lambda_1) x = \frac{\nu}{\tau^2} x$$

Substitute $\lambda_2 = \frac{1}{2\tau^2}$:

$$\left(\frac{\mu}{\tau^2} - \lambda_1\right) x = \frac{\nu}{\tau^2} x$$

This implies $\nu = \mu - \lambda_1 \tau^2$. However, we know from the mean constraint that $\mathbb{E}[X] = \mu$, so the mean of this distribution *must* be $\mu$. Therefore, $\nu = \mu$, which implies $\lambda_1 = 0$.

So, the distribution simplifies to:

$$p(x) = A \exp\left( -\lambda_2 (x-\mu)^2 \right)$$

Now, we use the remaining constraints to find $A$ and $\lambda_2$.

#### Using the Variance Constraint

$$\int_{-\infty}^{\infty} (x-\mu)^2 p(x) dx = \sigma^2$$

Substituting $p(x) = A \exp\left( -\lambda_2 (x-\mu)^2 \right)$:

$$A \int_{-\infty}^{\infty} (x-\mu)^2 e^{-\lambda_2 (x-\mu)^2} dx = \sigma^2$$

This integral is related to the second moment of a Gaussian. For a Gaussian distribution $N(\mu, \sigma_{eff}^2)$, the variance is $\sigma_{eff}^2$. Comparing the exponent $- \lambda_2 (x-\mu)^2$ with $- \frac{(x-\mu)^2}{2\sigma_{eff}^2}$, we get $\lambda_2 = \frac{1}{2\sigma_{eff}^2}$.

From the constraint, we know the variance is $\sigma^2$, so $\sigma_{eff}^2 = \sigma^2$. Therefore:

$$\lambda_2 = \frac{1}{2\sigma^2}$$

#### Using the Normalization Constraint

$$\int_{-\infty}^{\infty} p(x) dx = 1$$

$$A \int_{-\infty}^{\infty} e^{-\lambda_2 (x-\mu)^2} dx = 1$$

This is a standard Gaussian integral. We know that $\int_{-\infty}^{\infty} e^{-ax^2} dx = \sqrt{\frac{\pi}{a}}$.

Here, $a = \lambda_2 = \frac{1}{2\sigma^2}$. So,

$$A \sqrt{\frac{\pi}{\lambda_2}} = 1$$

$$A \sqrt{2\pi\sigma^2} = 1$$

$$A = \frac{1}{\sqrt{2\pi\sigma^2}}$$

### 6. The Maximum Entropy Distribution

Substituting the values of the Lagrange multipliers back into the form of $p(x)$:

*   $A = \frac{1}{\sqrt{2\pi\sigma^2}}$
*   $\lambda_1 = 0$
*   $\lambda_2 = \frac{1}{2\sigma^2}$

We get:

$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{1}{2\sigma^2} (x-\mu)^2 \right)$$

This is the probability density function of a Gaussian (Normal) distribution with mean $\mu$ and variance $\sigma^2$.

Therefore, the law of $X$ that maximizes entropy, given its mean $\mu$ and variance $\sigma^2$, is the Gaussian distribution:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

Let $P_{y, \mathbf{x}}$ denote $P(Y=y, \mathbf{x})$.

#### a. Define Entropy
The entropy for a discrete joint probability distribution is given by:
$$H(P) = -\sum_{y \in \{0,1\}} \sum_{\mathbf{x} \in \mathscr{X}} P_{y, \mathbf{x}} \ln P_{y, \mathbf{x}}$$
Our objective is to maximize $H(P)$.

#### b. Define Constraints

1.  **Normalization Constraint:** The sum of all probabilities must be 1.
    $$\sum_{y \in \{0,1\}} \sum_{\mathbf{x} \in \mathscr{X}} P_{y, \mathbf{x}} = 1$$

2.  **Marginal Probability Constraints for $\mathbf{x}$ (given scenarios):**
    *   $P(\mathbf{x}_A) = P_{0, (1,0,0)} + P_{1, (1,0,0)} = 0.40$
    *   $P(\mathbf{x}_B) = P_{0, (0,1,0)} + P_{1, (0,1,0)} = 0.25$
    *   $P(\mathbf{x}_C) = P_{0, (0,0,1)} + P_{1, (0,0,1)} = 0.05$
    *   $P(\mathbf{x}_D) = P_{0, (0,1,1)} + P_{1, (0,1,1)} = 0.15$
    *   $P(\mathbf{x}_E) = P_{0, (1,1,0)} + P_{1, (1,1,0)} = 0.05$
    For the unlisted states, we know $P(\mathbf{x}_F) + P(\mathbf{x}_G) + P(\mathbf{x}_H) = 0.10$. We'll handle this in the numerical solution by assigning $P(\mathbf{x}_F) = P(\mathbf{x}_G) = P(\mathbf{x}_H) = 0.10/3$.

3.  **Joint Expectation Constraints (for $Y=1$):**
    *   Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \sum_{\mathbf{x} \in \mathscr{X}} x_1 P_{1, \mathbf{x}} = 0.05$
        This expands to: $P_{1, (1,0,0)} + P_{1, (1,1,0)} + P_{1, (1,0,1)} + P_{1, (1,1,1)} = 0.05$

    *   Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \sum_{\mathbf{x} \in \mathscr{X}} x_2 P_{1, \mathbf{x}} = 0.15$
        This expands to: $P_{1, (0,1,0)} + P_{1, (1,1,0)} + P_{1, (0,1,1)} + P_{1, (1,1,1)} = 0.15$

    *   Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \sum_{\mathbf{x} \in \mathscr{X}} x_3 P_{1, \mathbf{x}} = 0.16$
        This expands to: $P_{1, (0,0,1)} + P_{1, (1,0,1)} + P_{1, (0,1,1)} + P_{1, (1,1,1)} = 0.16$

#### c. Set up the Lagrangian
To maximize the entropy subject to these constraints, we introduce Lagrange multipliers. Let $\lambda_0$ be for normalization, $\mu_{\mathbf{x}}$ for each marginal $P(\mathbf{x})$, and $\nu_1, \nu_2, \nu_3$ for the joint expectation constraints. The Lagrangian $L$ is:

$$L = -\sum_{y, \mathbf{x}} P_{y, \mathbf{x}} \ln P_{y, \mathbf{x}} - \lambda_0 \left(\sum_{y, \mathbf{x}} P_{y, \mathbf{x}} - 1\right) - \sum_{\mathbf{x}' \in \mathscr{X}} \mu_{\mathbf{x}'} \left(\sum_y P_{y, \mathbf{x}'} - P(\mathbf{x}')\right)$$
$$- \nu_1 \left(\sum_{\mathbf{x}} x_1 P_{1, \mathbf{x}} - 0.05\right) - \nu_2 \left(\sum_{\mathbf{x}} x_2 P_{1, \mathbf{x}} - 0.15\right) - \nu_3 \left(\sum_{\mathbf{x}} x_3 P_{1, \mathbf{x}} - 0.16\right)$$

#### d. Derive the Form of the Solution
Taking the derivative of $L$ with respect to each $P_{y, \mathbf{x}}$ and setting it to zero:

$$\frac{\partial L}{\partial P_{y, \mathbf{x}}} = -\ln P_{y, \mathbf{x}} - 1 - \lambda_0 - \mu_{\mathbf{x}} - (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \mathbb{I}(y=1) = 0$$

Rearranging for $P_{y, \mathbf{x}}$:

$$P_{y, \mathbf{x}} = \exp\left(-1 - \lambda_0 - \mu_{\mathbf{x}} - (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \mathbb{I}(y=1)\right)$$

Let $A_{\mathbf{x}} = \exp(-1 - \lambda_0 - \mu_{\mathbf{x}})$. Then:

$$P_{y, \mathbf{x}} = A_{\mathbf{x}} \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \mathbb{I}(y=1)\right)$$

Now, sum over $y$ to relate $A_{\mathbf{x}}$ to $P(\mathbf{x})$:

$$P(\mathbf{x}) = \sum_y P_{y, \mathbf{x}} = A_{\mathbf{x}} \left( \exp(0) + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right) \right)$$
$$P(\mathbf{x}) = A_{\mathbf{x}} \left( 1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right) \right)$$

From this, we can express $A_{\mathbf{x}}$:

$$A_{\mathbf{x}} = \frac{P(\mathbf{x})}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}$$

Substituting $A_{\mathbf{x}}$ back into the expression for $P_{y, \mathbf{x}}$:

*   For $y=0$ (Finance):
    $$P_{0, \mathbf{x}} = P(\mathbf{x}) \frac{1}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}$$

*   For $y=1$ (Technical):
    $$P_{1, \mathbf{x}} = P(\mathbf{x}) \frac{\exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}$$

This is the form of a log-linear model, where $P(Y=1|\mathbf{x}) = \frac{\exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}$ is the logistic sigmoid function. The Lagrange multipliers $\nu_1, \nu_2, \nu_3$ act as weights for the features.

#### e. System of Equations for Lagrange Multipliers
To find the specific values of $\nu_1, \nu_2, \nu_3$, we must substitute the derived forms of $P_{1, \mathbf{x}}$ back into the joint expectation constraints:

1.  **Constraint 1:**
    $$\sum_{\mathbf{x} \in \mathscr{X}} x_1 P(\mathbf{x}) \frac{\exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)} = 0.05$$

2.  **Constraint 2:**
    $$\sum_{\mathbf{x} \in \mathscr{X}} x_2 P(\mathbf{x}) \frac{\exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)} = 0.15$$

3.  **Constraint 3:**
    $$\sum_{\mathbf{x} \in \mathscr{X}} x_3 P(\mathbf{x}) \frac{\exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)}{1 + \exp\left(- (\nu_1 x_1 + \nu_2 x_2 + \nu_3 x_3) \right)} = 0.16$$

This is a system of three non-linear equations with three unknowns (the Lagrange multipliers $\nu_1, \nu_2, \nu_3$). This system does not generally have an analytical closed-form solution and must be solved numerically.


### 2. Numerical Solution for Lagrange Multipliers

We will use `scipy.optimize.fsolve` to find the roots of the system of non-linear equations derived above. This will give us the values for $\nu_1, \nu_2, \nu_3$. First, let's define the various $\mathbf{x}$ states and their marginal probabilities.

```python
import numpy as np
from scipy.optimize import fsolve

# Define all 8 possible x states (binary vectors for x1, x2, x3)
x_states = [
    (0, 0, 0),  # F
    (1, 0, 0),  # A
    (0, 1, 0),  # B
    (0, 0, 1),  # C
    (1, 1, 0),  # E
    (1, 0, 1),  # G
    (0, 1, 1),  # D
    (1, 1, 1)   # H
]

# Corresponding marginal probabilities P(x)
# Sum of given probabilities: 0.40 + 0.25 + 0.05 + 0.15 + 0.05 = 0.90
# Remaining 0.10 for states (0,0,0), (1,0,1), (1,1,1)
# Distribute uniformly among the unlisted states for MaxEnt principle
p_x_marginals = {
    (0, 0, 0): 0.10 / 3, # Assumed
    (1, 0, 0): 0.40,     # Scenario A
    (0, 1, 0): 0.25,     # Scenario B
    (0, 0, 1): 0.05,     # Scenario C
    (1, 1, 0): 0.05,     # Scenario E
    (1, 0, 1): 0.10 / 3, # Assumed
    (0, 1, 1): 0.15,     # Scenario D
    (1, 1, 1): 0.10 / 3  # Assumed
}

# Target expectation constraints for Y=1
target_exp_c1 = 0.05 # E[x1 * I(Y=1)]
target_exp_c2 = 0.15 # E[x2 * I(Y=1)]
target_exp_c3 = 0.16 # E[x3 * I(Y=1)]

# Define the system of equations to solve for nu1, nu2, nu3
def equations(nu_params):
    nu1, nu2, nu3 = nu_params

    # Calculate the current expectations based on current nu_params
    current_exp_c1 = 0.0
    current_exp_c2 = 0.0
    current_exp_c3 = 0.0

    for x_tuple in x_states:
        x1, x2, x3 = x_tuple
        P_x = p_x_marginals[x_tuple]

        # Calculate the argument for the exponential
        eta_x = - (nu1 * x1 + nu2 * x2 + nu3 * x3)

        # Calculate P(Y=1 | x) using the sigmoid function
        # Note: 1 / (1 + exp(-z)) = exp(z) / (1 + exp(z)) - this is the standard logistic form
        # Our derivation used exp(-eta_x) in the numerator
        # P(Y=1|x) = exp(eta_x) / (1 + exp(eta_x))
        # But the problem is max entropy formulation, with parameter signs such that
        # P_1,x = P(x) * exp(-(nu1*x1 + ...)) / (1 + exp(-(nu1*x1 + ...)))
        # So, the exponent is already negative as per our derivation.

        denominator = 1 + np.exp(eta_x)
        p_y1_given_x = np.exp(eta_x) / denominator

        # Contribution to expectations
        current_exp_c1 += x1 * P_x * p_y1_given_x
        current_exp_c2 += x2 * P_x * p_y1_given_x
        current_exp_c3 += x3 * P_x * p_y1_given_x

    # The system of equations is (current_exp - target_exp) = 0
    eq1 = current_exp_c1 - target_exp_c1
    eq2 = current_exp_c2 - target_exp_c2
    eq3 = current_exp_c3 - target_exp_c3

    return [eq1, eq2, eq3]

# Initial guess for nu1, nu2, nu3
# A common starting point is zeros, or small random values.
initial_guess = [0.0, 0.0, 0.0]

# Solve the system of equations
nu_solution, info, ier, msg = fsolve(equations, initial_guess, full_output=True)

if ier == 1:
    nu1_sol, nu2_sol, nu3_sol = nu_solution
    print(f"Successfully found Lagrange multipliers:")
    print(f"  nu1 = {nu1_sol:.4f}")
    print(f"  nu2 = {nu2_sol:.4f}")
    print(f"  nu3 = {nu3_sol:.4f}")

    # Now, calculate the resulting P(Y=y, x) for each state
    print("\nCalculated P(Y=y, x) distribution:")
    results = []
    for x_tuple in x_states:
        x1, x2, x3 = x_tuple
        P_x = p_x_marginals[x_tuple]

        eta_x = - (nu1_sol * x1 + nu2_sol * x2 + nu3_sol * x3)

        # P(Y=1|x) is the sigmoid of eta_x
        p_y1_given_x = np.exp(eta_x) / (1 + np.exp(eta_x))
        p_y0_given_x = 1 - p_y1_given_x

        P_y1_x = P_x * p_y1_given_x
        P_y0_x = P_x * p_y0_given_x

        results.append({
            'x': x_tuple,
            'P(x)': P_x,
            'P(Y=0|x)': p_y0_given_x,
            'P(Y=1|x)': p_y1_given_x,
            'P(Y=0, x)': P_y0_x,
            'P(Y=1, x)': P_y1_x
        })

    # Display results in a structured way (e.g., using pandas DataFrame)
    import pandas as pd
    df_results = pd.DataFrame(results)
    display(df_results)

    print(f"\nTotal sum of P(Y, x): {df_results['P(Y=0, x)'].sum() + df_results['P(Y=1, x)'].sum():.4f}")

    # Verify constraints
    calculated_exp_c1 = sum(r['x'][0] * r['P(Y=1, x)'] for r in results)
    calculated_exp_c2 = sum(r['x'][1] * r['P(Y=1, x)'] for r in results)
    calculated_exp_c3 = sum(r['x'][2] * r['P(Y=1, x)'] for r in results)

    print(f"\nVerification of constraints (should match targets):")
    print(f"  E[x1 * I(Y=1)]: Calculated = {calculated_exp_c1:.4f}, Target = {target_exp_c1}")
    print(f"  E[x2 * I(Y=1)]: Calculated = {calculated_exp_c2:.4f}, Target = {target_exp_c2}")
    print(f"  E[x3 * I(Y=1)]: Calculated = {calculated_exp_c3:.4f}, Target = {target_exp_c3}")

else:
    print(f"Failed to find Lagrange multipliers: {msg}")

```

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



### 1. Problem Setup: Maximum Entropy Formulation for Asset Price

We aim to find the continuous probability density function $p(S_T)$ for $S_T > 0$ that maximizes entropy subject to the given market constraints.

#### a. Define Continuous Entropy (Differential Entropy)
The differential entropy for a continuous random variable $S_T$ with probability density function $p(S_T)$ is given by:
$$H(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) dS_T$$
Our objective is to maximize this quantity.

#### b. Define Constraints
We have four constraints based on the available market data:

1.  **Normalization Constraint:** The total probability must integrate to 1.
    $$\int_{0}^{\infty} p(S_T) dS_T = 1$$

2.  **Mean Constraint (Risk-Free Growth):** The expected future price is given.
    $$\mathbb{E}[S_T] = \int_{0}^{\infty} S_T p(S_T) dS_T = 102$$

3.  **Second Moment Constraint (Market Volatility):** The expected squared price is given.
    $$\mathbb{E}[S_T^2] = \int_{0}^{\infty} S_T^2 p(S_T) dS_T = 10600$$

4.  **Logarithmic Expectation Constraint (Asymmetric Market Sentiment/Skew):** The expected value of $\ln(S_T)$ is given.
    $$\mathbb{E}[\ln(S_T)] = \int_{0}^{\infty} \ln(S_T) p(S_T) dS_T = 4.60$$

#### c. Set up the Lagrangian
To maximize the differential entropy subject to these constraints, we use the method of Lagrange multipliers. Let $\lambda_0, \lambda_1, \lambda_2, \lambda_3$ be the Lagrange multipliers corresponding to the normalization, mean, second moment, and logarithmic expectation constraints, respectively. The Lagrangian $L$ is:

$$L = -\int_{0}^{\infty} p(S_T) \ln p(S_T) dS_T \\ - \lambda_0 \left(\int_{0}^{\infty} p(S_T) dS_T - 1\right) \\ - \lambda_1 \left(\int_{0}^{\infty} S_T p(S_T) dS_T - 102\right) \\ - \lambda_2 \left(\int_{0}^{\infty} S_T^2 p(S_T) dS_T - 10600\right) \\ - \lambda_3 \left(\int_{0}^{\infty} \ln(S_T) p(S_T) dS_T - 4.60\right)$$

#### d. Derive the Form of the Solution
To find the function $p(S_T)$ that maximizes $L$, we take the variational derivative with respect to $p(S_T)$ and set it to zero:

$$\frac{\delta L}{\delta p(S_T)} = -(\ln p(S_T) + 1) - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$

Rearranging the terms to solve for $\ln p(S_T)$:

$$\ln p(S_T) = -1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)$$

Exponentiating both sides, we get the form of $p(S_T)$:

$$p(S_T) = \exp\left(-1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)\right)$$

We can rewrite this as:

$$p(S_T) = \exp(-1 - \lambda_0) \exp(-\lambda_1 S_T) \exp(-\lambda_2 S_T^2) \exp(-\lambda_3 \ln(S_T))$$

Using the property $e^{a \ln(x)} = x^a$, we get:

$$p(S_T) = C \cdot S_T^{-\lambda_3} \exp(-\lambda_1 S_T - \lambda_2 S_T^2)$$

where $C = \exp(-1 - \lambda_0)$ is a normalization constant. The parameters $\lambda_1, \lambda_2, \lambda_3$ (and implicitly $C$) must be determined by satisfying the constraints. For the integral to converge over $(0, \infty)$, we expect $\lambda_2 > 0$, and for $S_T^{-\lambda_3}$ not to cause divergence at $S_T=0$, we expect $-\lambda_3 > -1$, or $\lambda_3 < 1$. This probability density function is a non-standard distribution, and the Lagrange multipliers must be found numerically.


### 2. Numerical Solution for Lagrange Multipliers

To find the specific values of $\lambda_1, \lambda_2, \lambda_3$ (and the normalization constant $C$), we need to solve a system of non-linear equations derived from the constraints. This will involve numerical integration, as there are no elementary analytical solutions for the integrals involved. We will use `scipy.integrate.quad` for numerical integration and `scipy.optimize.fsolve` to find the roots of the system of equations.

The probability density function is $p(S_T) = C \cdot S_T^{-\lambda_3} \exp(-\lambda_1 S_T - \lambda_2 S_T^2)$. The normalization constant $C$ is given by $C = 1 / \int_{0}^{\infty} S_T^{-\lambda_3} \exp(-\lambda_1 S_T - \lambda_2 S_T^2) dS_T$. We will define a function that takes the Lagrange multipliers as input, calculates $C$, and then calculates the expected values for each constraint. The goal is to make these calculated expectations match the target values.


```python
import numpy as np
from scipy.optimize import fsolve
from scipy.integrate import quad

# Define the target constraint values
TARGET_MEAN = 102
TARGET_SECOND_MOMENT = 10600
TARGET_LOG_EXPECTATION = 4.60

# Define the unnormalized probability density function component
def unnormalized_pdf_component(s_t, lambda1, lambda2, lambda3):
    # Ensure s_t > 0 for ln(s_t) and s_t**(-lambda3)
    if s_t <= 0:
        return 0.0 # Or a very small number for numerical stability if integration needs it, but quad handles limits.

    # For the integral to converge, lambda2 should be positive and lambda3 less than 1
    # if lambda3 >= 1, s_t**(-lambda3) might cause issues near 0
    # We set a small lower bound for integration to avoid division by zero for s_t**(-lambda3)

    return (s_t**(-lambda3)) * np.exp(-lambda1 * s_t - lambda2 * s_t**2)

# Define the system of equations to solve for lambda1, lambda2, lambda3
def equations(lambdas):
    lambda1, lambda2, lambda3 = lambdas

    # Constraints on lambda values for numerical stability and integral convergence
    # lambda2 must be positive for exp(-lambda2 * s_t**2) to ensure convergence at infinity
    # lambda3 must be less than 1 for s_t**(-lambda3) to ensure convergence at 0
    if lambda2 <= 0 or lambda3 >= 1: # or other conditions based on integrand behavior
        return [1e10, 1e10, 1e10] # Return large errors to guide solver away from these regions

    # 1. Calculate the normalization constant Z (partition function)
    # The lower limit of integration for quad should be > 0 due to s_t**(-lambda3)
    # A very small epsilon is used to avoid issues at 0.
    integral_Z, _ = quad(unnormalized_pdf_component, 1e-10, np.inf, args=(lambda1, lambda2, lambda3))

    if integral_Z <= 0: # Ensure positive probability mass
        return [1e10, 1e10, 1e10]

    C = 1.0 / integral_Z

    # Define the actual pdf
    def pdf(s_t):
        return C * unnormalized_pdf_component(s_t, lambda1, lambda2, lambda3)

    # 2. Calculate the current mean
    integrand_mean = lambda s_t: s_t * pdf(s_t)
    current_mean, _ = quad(integrand_mean, 1e-10, np.inf)

    # 3. Calculate the current second moment
    integrand_second_moment = lambda s_t: (s_t**2) * pdf(s_t)
    current_second_moment, _ = quad(integrand_second_moment, 1e-10, np.inf)

    # 4. Calculate the current logarithmic expectation
    integrand_log_expectation = lambda s_t: np.log(s_t) * pdf(s_t)
    current_log_expectation, _ = quad(integrand_log_expectation, 1e-10, np.inf)

    # Return the differences (errors) for fsolve to minimize
    eq1 = current_mean - TARGET_MEAN
    eq2 = current_second_moment - TARGET_SECOND_MOMENT
    eq3 = current_log_expectation - TARGET_LOG_EXPECTATION

    return [eq1, eq2, eq3]

# Initial guess for lambda1, lambda2, lambda3
# This is often the trickiest part for fsolve.
# Based on the functional form, small positive values for lambda1, lambda2,
# and a value < 1 for lambda3 might be a good starting point.
# A value of lambda3 near 0 would make it closer to a modified Gaussian.
initial_guess = [0.01, 0.0001, 0.5] # Example guess, may need adjustment

print("Attempting to solve for Lagrange multipliers...")

# Solve the system of equations
try:
    # Use a different solver or parameters if fsolve struggles
    # 'xtol' is the relative error desired for the solution
    lambda_solution = fsolve(equations, initial_guess, xtol=1e-8, maxfev=2000)

    lambda1_sol, lambda2_sol, lambda3_sol = lambda_solution
    print(f"Successfully found Lagrange multipliers:")
    print(f"  lambda1 = {lambda1_sol:.6f}")
    print(f"  lambda2 = {lambda2_sol:.6f}")
    print(f"  lambda3 = {lambda3_sol:.6f}")

    # Now, calculate the normalization constant with the found lambdas
    integral_Z_final, _ = quad(unnormalized_pdf_component, 1e-10, np.inf, args=(lambda1_sol, lambda2_sol, lambda3_sol))
    C_final = 1.0 / integral_Z_final

    print(f"\nNormalization Constant C = {C_final:.6e}")

    # Verify constraints with the obtained parameters
    def final_pdf(s_t):
        return C_final * unnormalized_pdf_component(s_t, lambda1_sol, lambda2_sol, lambda3_sol)

    calculated_mean, _ = quad(lambda s_t: s_t * final_pdf(s_t), 1e-10, np.inf)
    calculated_second_moment, _ = quad(lambda s_t: (s_t**2) * final_pdf(s_t), 1e-10, np.inf)
    calculated_log_expectation, _ = quad(lambda s_t: np.log(s_t) * final_pdf(s_t), 1e-10, np.inf)

    print(f"\nVerification of constraints (should match targets):")
    print(f"  Calculated Mean: {calculated_mean:.4f}, Target: {TARGET_MEAN}")
    print(f"  Calculated Second Moment: {calculated_second_moment:.4f}, Target: {TARGET_SECOND_MOMENT}")
    print(f"  Calculated Log Expectation: {calculated_log_expectation:.4f}, Target: {TARGET_LOG_EXPECTATION}")

    # Optionally, plot the PDF
    import matplotlib.pyplot as plt
    s_t_values = np.linspace(1, 200, 500) # Range for plotting
    pdf_values = [final_pdf(s) for s in s_t_values]

    fig = plt.figure(figsize=(10, 6))
    plt.plot(s_t_values, pdf_values)
    plt.title('Maximum Entropy Probability Density Function for Asset X')
    plt.xlabel('$S_T$ (Asset Price)')
    plt.ylabel('$p(S_T)$')
    plt.grid(True)
    plt.show()

except Exception as e:
    print(f"An error occurred during solving: {e}")
    print("Try adjusting the initial_guess or bounds for the Lagrange multipliers if the solver fails to converge.")
```
