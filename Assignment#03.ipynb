### Setup: Define the sample space, probabilities, and random variables

```python
import numpy as np

# Define the sample space and probabilities
omega = ['omega_1', 'omega_2', 'omega_3', 'omega_4']
p_omega = {
    'omega_1': 0.50,
    'omega_2': 0.10,
    'omega_3': 0.20,
    'omega_4': 0.20
}

# Define the random variable X (fraud indicator)
X = {
    'omega_1': 0,
    'omega_2': 1,
    'omega_3': 0,
    'omega_4': 1
}

# Define the random variable Y (customer risk class)
Y = {
    'omega_1': 0,
    'omega_2': 0,
    'omega_3': 1,
    'omega_4': 1
}

print("Sample Space (Ω):", omega)
print("Probabilities P(ω):", p_omega)
print("Fraud Indicator X(ω):", X)
print("Customer Risk Y(ω):", Y)
```

### 1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

The probability space $(\Omega, \mathfrak F, P)$ is defined as:

*   **Sample Space ($\Omega$):** The set of all possible outcomes.
    $\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$

*   **Sigma-algebra ($\mathfrak F$):** The set of all possible events. Since $\mathfrak F = \mathfrak P(\Omega)$ (the power set of $\Omega$), it includes all $2^4 = 16$ possible subsets of $\Omega$.
    For example, elements of $\mathfrak F$ include $\emptyset$, $\{\omega_1\}$, $\{\omega_1, \omega_2\}$, $\{\omega_1, \omega_2, \omega_3\}$, and $\Omega$ itself.

*   **Probability Measure ($P$):** A function that assigns a probability to each event in $\mathfrak F$. For singletons (elementary events), the probabilities are given as:
    $P(\{\omega_1\})=0.50$
    $P(\{\omega_2\})=0.10$
    $P(\{\omega_3\})=0.20$
    $P(\{\omega_4\})=0.20$

    For any other event $A \in \mathfrak F$, its probability is the sum of the probabilities of the elementary outcomes it contains. For example, $P(\{\omega_1, \omega_2\}) = P(\{\omega_1\}) + P(\{\omega_2\}) = 0.50 + 0.10 = 0.60$.

### 2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

```python
# Y maps outcomes to values:
# Y(ω1) = 0
# Y(ω2) = 0
# Y(ω3) = 1
# Y(ω4) = 1

# The pre-images of the values Y can take are:
# Y=0: {ω1, ω2}
# Y=1: {ω3, ω4}

# The σ-algebra generated by Y, σ(Y), consists of the empty set, the sample space,
# and all possible unions and complements of the pre-images of the values Y can take.

# The distinct values of Y are {0, 1}.
# The sets generated by Y are the pre-images of these values:
# Y^{-1}(0) = {ω ∈ Ω : Y(ω) = 0} = {ω1, ω2}
# Y^{-1}(1) = {ω ∈ Ω : Y(ω) = 1} = {ω3, ω4}

# σ(Y) is the smallest σ-algebra containing these sets.
# It must contain:
# 1. The empty set: ∅
# 2. The entire sample space: Ω = {ω1, ω2, ω3, ω4}
# 3. The pre-images themselves: {ω1, ω2} and {ω3, ω4}
# 4. Complements of these sets:
#    Complement of {ω1, ω2} is {ω3, ω4}
#    Complement of {ω3, ω4} is {ω1, ω2}
# 5. Unions of these sets:
#    {ω1, ω2} ∪ {ω3, ω4} = Ω
#    {ω1, ω2} ∪ ∅ = {ω1, ω2}
#    {ω3, ω4} ∪ ∅ = {ω3, ω4}

sigma_Y = [
    set(), # Empty set
    {'omega_1', 'omega_2'},
    {'omega_3', 'omega_4'},
    {'omega_1', 'omega_2', 'omega_3', 'omega_4'} # Omega
]

print("The σ-algebra generated by Y, σ(Y), is:")
for s in sigma_Y:
    print(s)
```

### 3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

The conditional expectation $E[X\mid Y]$ is defined as a random variable that is $\sigma(Y)$-measurable and satisfies the defining property of conditional expectation.

**What does $\sigma(Y)$-measurable mean?**
A random variable $Z$ is $\sigma(Y)$-measurable if for every Borel set $B$ (or more generally, any set in the sigma-algebra of the target space), the pre-image $Z^{-1}(B) = \{\omega \in \Omega : Z(\omega) \in B\}$ is an element of $\sigma(Y)$.

In simpler terms, if a random variable is $\sigma(Y)$-measurable, its value can only be determined by knowing the value of $Y$. It cannot distinguish between outcomes that have the same $Y$ value.

From the previous step, we found that the atoms of $\sigma(Y)$ (the smallest non-empty sets in $\sigma(Y)$ that cannot be further decomposed) are:
*   $A_0 = \{\omega_1, \omega_2\}$ (where $Y=0$)
*   $A_1 = \{\omega_3, \omega_4\}$ (where $Y=1$)

For $E[X\mid Y]$ to be $\sigma(Y)$-measurable, it must be constant on these atoms. If it were not constant on, say, $A_0$, then $E[X\mid Y](\omega_1)$ would be different from $E[X\mid Y](\omega_2)$, even though $Y(\omega_1) = Y(\omega_2) = 0$. This would mean $E[X\mid Y]$ could distinguish between $\omega_1$ and $\omega_2$, which are indistinguishable by $Y$ (and thus by $\sigma(Y)$).

Therefore, $E[X\mid Y]$ must take a single constant value for all $\omega \in \{\omega_1, \omega_2\}$ (since $Y$ is constant on this set), and another single constant value for all $\omega \in \{\omega_3, \omega_4\}$ (since $Y$ is constant on this set).

### 4. Compute $E[X\mid Y]$ explicitly and show that $E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}$

```python
# We need to calculate E[X | Y=0] and E[X | Y=1]

# For Y=0, the outcomes are {ω1, ω2}
# P(Y=0) = P({ω1}) + P({ω2}) = 0.50 + 0.10 = 0.60
# E[X | Y=0] = (X(ω1)P({ω1}) + X(ω2)P({ω2})) / P(Y=0)
#            = (0 * 0.50 + 1 * 0.10) / 0.60
#            = 0.10 / 0.60 = 1/6

# For Y=1, the outcomes are {ω3, ω4}
# P(Y=1) = P({ω3}) + P({ω4}) = 0.20 + 0.20 = 0.40
# E[X | Y=1] = (X(ω3)P({ω3}) + X(ω4)P({ω4})) / P(Y=1)
#            = (0 * 0.20 + 1 * 0.20) / 0.40
#            = 0.20 / 0.40 = 1/2

# So, E[X | Y] is a random variable that takes value 1/6 when Y=0 and 1/2 when Y=1.
# This can be written as:
# E[X | Y] = (1/6) * Indicator_{Y=0} + (1/2) * Indicator_{Y=1}

# Let's verify with code:

# Calculate P(Y=0) and P(Y=1)
p_Y_0 = p_omega['omega_1'] + p_omega['omega_2']
p_Y_1 = p_omega['omega_3'] + p_omega['omega_4']

print(f"P(Y=0) = {p_Y_0}")
print(f"P(Y=1) = {p_Y_1}")

# Calculate E[X | Y=0]
expected_X_given_Y_0 = (X['omega_1'] * p_omega['omega_1'] + X['omega_2'] * p_omega['omega_2']) / p_Y_0
print(f"E[X | Y=0] = ({X['omega_1']} * {p_omega['omega_1']} + {X['omega_2']} * {p_omega['omega_2']}) / {p_Y_0} = {expected_X_given_Y_0:.4f} (which is 1/6)")

# Calculate E[X | Y=1]
expected_X_given_Y_1 = (X['omega_3'] * p_omega['omega_3'] + X['omega_4'] * p_omega['omega_4']) / p_Y_1
print(f"E[X | Y=1] = ({X['omega_3']} * {p_omega['omega_3']} + {X['omega_4']} * {p_omega['omega_4']}) / {p_Y_1} = {expected_X_given_Y_1:.4f} (which is 1/2)")

# Construct E[X|Y] as a dictionary mapping omega to its value
E_X_given_Y = {}
for outcome in omega:
    if Y[outcome] == 0:
        E_X_given_Y[outcome] = expected_X_given_Y_0
    else:
        E_X_given_Y[outcome] = expected_X_given_Y_1

print("\nE[X | Y] values for each omega:", E_X_given_Y)
```

### 5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$, $\int_A E[X\mid Y]\,dP=\int_A X\,dP$.

```python
# The defining property of conditional expectation states that for any A in σ(Y):
# E[X * Indicator_A] = E[E[X|Y] * Indicator_A]
# In our discrete case, this means:
# Sum_{ω in A} X(ω)P({ω}) = Sum_{ω in A} E[X|Y](ω)P({ω})

# The sets A in σ(Y) are: ∅, {ω1, ω2}, {ω3, ω4}, Ω
# We only need to check the non-empty, non-Ω sets, as the others follow.

# Case 1: A = {ω1, ω2} (where Y=0)
print("Checking for A = {ω1, ω2} (where Y=0):")
integral_X_A = X['omega_1'] * p_omega['omega_1'] + X['omega_2'] * p_omega['omega_2']
integral_E_X_given_Y_A = E_X_given_Y['omega_1'] * p_omega['omega_1'] + E_X_given_Y['omega_2'] * p_omega['omega_2']

print(f"  ∫_A X dP = {integral_X_A:.4f}")
print(f"  ∫_A E[X|Y] dP = {integral_E_X_given_Y_A:.4f}")
print(f"  Are they equal? {np.isclose(integral_X_A, integral_E_X_given_Y_A)}")

# Case 2: A = {ω3, ω4} (where Y=1)
print("\nChecking for A = {ω3, ω4} (where Y=1):")
integral_X_A = X['omega_3'] * p_omega['omega_3'] + X['omega_4'] * p_omega['omega_4']
integral_E_X_given_Y_A = E_X_given_Y['omega_3'] * p_omega['omega_3'] + E_X_given_Y['omega_4'] * p_omega['omega_4']

print(f"  ∫_A X dP = {integral_X_A:.4f}")
print(f"  ∫_A E[X|Y] dP = {integral_E_X_given_Y_A:.4f}")
print(f"  Are they equal? {np.isclose(integral_X_A, integral_E_X_given_Y_A)}")

# Case 3: A = Ω (the entire sample space)
print("\nChecking for A = Ω:")
integral_X_A = sum(X[o] * p_omega[o] for o in omega)
integral_E_X_given_Y_A = sum(E_X_given_Y[o] * p_omega[o] for o in omega)

print(f"  ∫_Ω X dP = {integral_X_A:.4f}")
print(f"  ∫_Ω E[X|Y] dP = {integral_E_X_given_Y_A:.4f}")
print(f"  Are they equal? {np.isclose(integral_X_A, integral_E_X_given_Y_A)}")
```

### 6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

```python
# The Conditional PMF p_{X|Y}(x | y) is P(X=x | Y=y) = P(X=x, Y=y) / P(Y=y)

# Joint probabilities P(X=x, Y=y)
# X takes values {0, 1}
# Y takes values {0, 1}

# P(X=0, Y=0): Outcomes where X=0 and Y=0 are {ω1}
# P(X=0, Y=0) = P({ω1}) = 0.50

# P(X=1, Y=0): Outcomes where X=1 and Y=0 are {ω2}
# P(X=1, Y=0) = P({ω2}) = 0.10

# P(X=0, Y=1): Outcomes where X=0 and Y=1 are {ω3}
# P(X=0, Y=1) = P({ω3}) = 0.20

# P(X=1, Y=1): Outcomes where X=1 and Y=1 are {ω4}
# P(X=1, Y=1) = P({ω4}) = 0.20

# From Q4, we have:
# P(Y=0) = 0.60
# P(Y=1) = 0.40

# Conditional PMF calculations:
print("Conditional PMF p_X|Y(x | y):\n")

# For y = 0:
px_given_y_0 = {}
px_given_y_0[0] = p_omega['omega_1'] / p_Y_0  # P(X=0, Y=0) / P(Y=0) = 0.50 / 0.60 = 5/6
px_given_y_0[1] = p_omega['omega_2'] / p_Y_0  # P(X=1, Y=0) / P(Y=0) = 0.10 / 0.60 = 1/6

print(f"p_X|Y(0 | 0) = P(X=0, Y=0) / P(Y=0) = {p_omega['omega_1']:.2f} / {p_Y_0:.2f} = {px_given_y_0[0]:.4f} (5/6)")
print(f"p_X|Y(1 | 0) = P(X=1, Y=0) / P(Y=0) = {p_omega['omega_2']:.2f} / {p_Y_0:.2f} = {px_given_y_0[1]:.4f} (1/6)")
print(f"  Sum for y=0: {px_given_y_0[0] + px_given_y_0[1]:.4f}\n")

# For y = 1:
px_given_y_1 = {}
px_given_y_1[0] = p_omega['omega_3'] / p_Y_1  # P(X=0, Y=1) / P(Y=1) = 0.20 / 0.40 = 1/2
px_given_y_1[1] = p_omega['omega_4'] / p_Y_1  # P(X=1, Y=1) / P(Y=1) = 0.20 / 0.40 = 1/2

print(f"p_X|Y(0 | 1) = P(X=0, Y=1) / P(Y=1) = {p_omega['omega_3']:.2f} / {p_Y_1:.2f} = {px_given_y_1[0]:.4f} (1/2)")
print(f"p_X|Y(1 | 1) = P(X=1, Y=1) / P(Y=1) = {p_omega['omega_4']:.2f} / {p_Y_1:.2f} = {px_given_y_1[1]:.4f} (1/2)")
print(f"  Sum for y=1: {px_given_y_1[0] + px_given_y_1[1]:.4f}\n")
```

### 7. Verify the identity: $\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$.

```python
# We will use the conditional PMF values calculated in the previous step.
# E[X | Y=y] = 0 * p_X|Y(0 | y) + 1 * p_X|Y(1 | y) = p_X|Y(1 | y)

print("Verifying the identity E[X | Y=y] = sum(x * p_X|Y(x | y)):\n")

# For y = 0:
sum_x_px_given_y_0 = 0 * px_given_y_0[0] + 1 * px_given_y_0[1]
print(f"For Y=0: sum(x * p_X|Y(x | 0)) = {sum_x_px_given_y_0:.4f}")
print(f"  From Q4, E[X | Y=0] = {expected_X_given_Y_0:.4f}")
print(f"  Are they equal? {np.isclose(sum_x_px_given_y_0, expected_X_given_Y_0)}\n")

# For y = 1:
sum_x_px_given_y_1 = 0 * px_given_y_1[0] + 1 * px_given_y_1[1]
print(f"For Y=1: sum(x * p_X|Y(x | 1)) = {sum_x_px_given_y_1:.4f}")
print(f"  From Q4, E[X | Y=1] = {expected_X_given_Y_1:.4f}")
print(f"  Are they equal? {np.isclose(sum_x_px_given_y_1, expected_X_given_Y_1)}\n")
```

### 8. Calculate the Mean Square Error (MSE)

```python
# The MSE for the conditional expectation E[X|Y] as an estimator for X is E[(X - E[X|Y])^2]
# This can be calculated as a sum over the sample space:
# MSE = Sum_{ω in Ω} (X(ω) - E[X|Y](ω))^2 * P({ω})

# Recall E_X_given_Y from Q4
# E[X|Y](ω1) = 1/6
# E[X|Y](ω2) = 1/6
# E[X|Y](ω3) = 1/2
# E[X|Y](ω4) = 1/2

mse = 0
for outcome in omega:
    error_squared = (X[outcome] - E_X_given_Y[outcome])**2
    mse += error_squared * p_omega[outcome]

print(f"Mean Square Error (MSE) = E[(X - E[X|Y])^2] = {mse:.4f}")

# Let's break it down:
# (X(ω1) - E[X|Y](ω1))^2 * P(ω1) = (0 - 1/6)^2 * 0.50 = (1/36) * 0.50 = 0.013888...
# (X(ω2) - E[X|Y](ω2))^2 * P(ω2) = (1 - 1/6)^2 * 0.10 = (5/6)^2 * 0.10 = (25/36) * 0.10 = 0.069444...
# (X(ω3) - E[X|Y](ω3))^2 * P(ω3) = (0 - 1/2)^2 * 0.20 = (1/4) * 0.20 = 0.05
# (X(ω4) - E[X|Y](ω4))^2 * P(ω4) = (1 - 1/2)^2 * 0.20 = (1/4) * 0.20 = 0.05
# Total = 0.013888 + 0.069444 + 0.05 + 0.05 = 0.188333...

calculated_mse = (0 - 1/6)**2 * 0.50 + \
                 (1 - 1/6)**2 * 0.10 + \
                 (0 - 1/2)**2 * 0.20 + \
                 (1 - 1/2)**2 * 0.20

print(f"Calculated MSE step-by-step = {calculated_mse:.4f} (which is 11/60 or 0.1833)")
```

### 9. How would the Mean Square Error (MSE) change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the Law of Total Variance.

```python
# First, let's calculate E[X]
E_X = sum(X[outcome] * p_omega[outcome] for outcome in omega)
print(f"E[X] = {E_X:.2f}")

# The problem states E[X] = 0.30, which matches our calculation.

# If the bank uses E[X] as a constant estimate for every transaction, the MSE will be E[(X - E[X])^2],
# which is simply the variance of X, Var(X).

mse_no_Y = 0
for outcome in omega:
    error_squared = (X[outcome] - E_X)**2
    mse_no_Y += error_squared * p_omega[outcome]

print(f"\nMSE if ignoring Y (Var(X)) = E[(X - E[X])^2] = {mse_no_Y:.4f}")

# Relationship to the Law of Total Variance:
# The Law of Total Variance states: Var(X) = E[Var(X|Y)] + Var(E[X|Y])
# Where E[Var(X|Y)] is the MSE we calculated in Q8: E[(X - E[X|Y])^2]
# And Var(E[X|Y]) is the variance of the conditional expectation itself.

# Let's calculate Var(E[X|Y])
# E[X|Y] takes value 1/6 with probability P(Y=0)=0.60
# E[X|Y] takes value 1/2 with probability P(Y=1)=0.40

# E[E[X|Y]] = E_X (this is a property of conditional expectation)
# Let's verify:
E_E_X_given_Y = (1/6) * p_Y_0 + (1/2) * p_Y_1
print(f"E[E[X|Y]] = {E_E_X_given_Y:.4f}. This should be equal to E[X] = {E_X:.2f}. (Verification: {np.isclose(E_E_X_given_Y, E_X)})\n")

var_E_X_given_Y = (1/6 - E_X)**2 * p_Y_0 + (1/2 - E_X)**2 * p_Y_1
print(f"Var(E[X|Y]) = {var_E_X_given_Y:.4f}")

# Now, let's check the Law of Total Variance:
# Var(X) = E[Var(X|Y)] + Var(E[X|Y])
# Var(X) = mse_no_Y
# E[Var(X|Y)] = mse (from Q8)

print(f"\nLaw of Total Variance verification:")
print(f"  Var(X) = {mse_no_Y:.4f}")
print(f"  E[Var(X|Y)] + Var(E[X|Y]) = {mse:.4f} + {var_E_X_given_Y:.4f} = {mse + var_E_X_given_Y:.4f}")
print(f"  Are they equal? {np.isclose(mse_no_Y, mse + var_E_X_given_Y)}")

print("\nExplanation:")
print("The MSE without considering Y is simply the variance of X. When we do consider Y, the MSE is E[Var(X|Y)].")
print("The Law of Total Variance shows that the total variance of X (the error when ignoring Y) can be decomposed into two parts:")
print("1. The expected conditional variance, E[Var(X|Y)], which is the MSE when using E[X|Y] as the estimator (our calculated MSE in Q8). This represents the irreducible error after considering Y.")
print("2. The variance of the conditional expectation, Var(E[X|Y]), which represents the reduction in variance (and thus error) gained by using Y. This is the information Y provides about X.")
print("Since Var(E[X|Y]) is always non-negative, the MSE when using E[X|Y] will always be less than or equal to the MSE when ignoring Y (Var(X)).")
print(f"In this case, the MSE decreased from {mse_no_Y:.4f} (without Y) to {mse:.4f} (with Y), showing a reduction in error of {var_E_X_given_Y:.4f} by incorporating the customer risk class Y.")
```

### 10. Using the joint distribution of $(X, Y)$ derived from the sample space:

```python
import math

# Joint probabilities P(X=x, Y=y) as derived in Q6:
# P(X=0, Y=0) = 0.50
# P(X=1, Y=0) = 0.10
# P(X=0, Y=1) = 0.20
# P(X=1, Y=1) = 0.20

# Marginal probabilities for X:
# P(X=0) = P(X=0, Y=0) + P(X=0, Y=1) = 0.50 + 0.20 = 0.70
# P(X=1) = P(X=1, Y=0) + P(X=1, Y=1) = 0.10 + 0.20 = 0.30

# Marginal probabilities for Y (from Q4):
# P(Y=0) = 0.60
# P(Y=1) = 0.40

# Conditional probabilities p_X|Y(x|y) from Q6:
# p_X|Y(0|0) = 5/6, p_X|Y(1|0) = 1/6
# p_X|Y(0|1) = 1/2, p_X|Y(1|1) = 1/2

p_X_0 = 0.70
p_X_1 = 0.30
p_Y_0 = 0.60
p_Y_1 = 0.40

px_given_y_0_0 = 5/6
px_given_y_1_0 = 1/6
px_given_y_0_1 = 1/2
px_given_y_1_1 = 1/2

# Helper function for entropy calculation (for discrete variables)
def entropy(probabilities):
    h = 0
    for p in probabilities:
        if p > 0:
            h -= p * math.log2(p)
    return h

# a) Compute the entropy H(X)
print("a) Compute the entropy H(X):")
H_X = entropy([p_X_0, p_X_1])
print(f"H(X) = {H_X:.4f} bits")

# b) Compute the conditional entropy H(X | Y)
print("\nb) Compute the conditional entropy H(X | Y):")
# H(X|Y) = P(Y=0)H(X|Y=0) + P(Y=1)H(X|Y=1)

# H(X|Y=0) = - (p_X|Y(0|0)log2(p_X|Y(0|0)) + p_X|Y(1|0)log2(p_X|Y(1|0)))
H_X_given_Y_0 = entropy([px_given_y_0_0, px_given_y_1_0])
print(f"  H(X|Y=0) = {H_X_given_Y_0:.4f} bits")

# H(X|Y=1) = - (p_X|Y(0|1)log2(p_X|Y(0|1)) + p_X|Y(1|1)log2(p_X|Y(1|1)))
H_X_given_Y_1 = entropy([px_given_y_0_1, px_given_y_1_1])
print(f"  H(X|Y=1) = {H_X_given_Y_1:.4f} bits")

H_X_given_Y = p_Y_0 * H_X_given_Y_0 + p_Y_1 * H_X_given_Y_1
print(f"H(X|Y) = P(Y=0)H(X|Y=0) + P(Y=1)H(X|Y=1) = {H_X_given_Y:.4f} bits")

# c) Calculate the Mutual Information I(X; Y)
print("\nc) Calculate the Mutual Information I(X; Y):")
# I(X;Y) = H(X) - H(X|Y)
I_X_Y = H_X - H_X_given_Y
print(f"I(X;Y) = H(X) - H(X|Y) = {I_X_Y:.4f} bits")

# d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?
print("\nd) Percentage of uncertainty removed by knowing the customer's risk class:")
percentage_removed = (I_X_Y / H_X) * 100
print(f"Percentage of uncertainty about X removed by knowing Y = (I(X;Y) / H(X)) * 100 = {percentage_removed:.2f}%")

print("\nInterpretation:")
print(f"Initially, the uncertainty about fraud (H(X)) is {H_X:.4f} bits. After knowing the customer's risk class (Y), the remaining uncertainty (H(X|Y)) is {H_X_given_Y:.4f} bits. The mutual information I(X;Y) = {I_X_Y:.4f} bits represents the amount of uncertainty reduction. This means {percentage_removed:.2f}% of the total uncertainty about fraud is removed by knowing whether the customer is low-risk or high-risk.")
```

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

### Derivation of the Conditional Distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ for a Multivariate Gaussian

To find the conditional PDF $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

We know the joint PDF $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ (where $\mathbf{z} = \begin{bmatrix} \mathbf{x} \\ \mathbf{y} \end{bmatrix}$) is:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right)$$

And the marginal PDF $f_{\mathbf{Y}}(\mathbf{y})$ is also a multivariate Gaussian with mean $\boldsymbol{\mu}_Y$ and covariance $\boldsymbol{\Sigma}_{YY}$:

$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)$$

The core of the derivation lies in carefully expanding the quadratic form in the exponent of the joint PDF. Let $\mathbf{z} - \boldsymbol{\mu} = \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_X \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix}$.

Let the inverse of the covariance matrix $\boldsymbol{\Sigma}^{-1}$ be partitioned similarly:

$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \boldsymbol{\Lambda}_{XX} & \boldsymbol{\Lambda}_{XY} \\ \boldsymbol{\Lambda}_{YX} & \boldsymbol{\Lambda}_{YY} \end{bmatrix}$$

Where $\boldsymbol{\Lambda}_{XX}$, $\boldsymbol{\Lambda}_{XY}$, etc., are sub-matrices of $\boldsymbol{\Sigma}^{-1}$. Using block matrix inversion formulas, these are related to the blocks of $\boldsymbol{\Sigma}$ as:

*   $\boldsymbol{\Lambda}_{XX} = (\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})^{-1} = \boldsymbol{\Sigma}_{X\mid Y}^{-1}$
*   $\boldsymbol{\Lambda}_{YY} = (\boldsymbol{\Sigma}_{YY} - \boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{XX}^{-1}\boldsymbol{\Sigma}_{XY})^{-1}$
*   $\boldsymbol{\Lambda}_{XY} = -\boldsymbol{\Sigma}_{X\mid Y}^{-1} \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$
*   $\boldsymbol{\Lambda}_{YX} = -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1}$

Now, expand the quadratic term in the exponent of the joint PDF:

$$ (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) = \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_X \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Lambda}_{XX} & \boldsymbol{\Lambda}_{XY} \\ \boldsymbol{\Lambda}_{YX} & \boldsymbol{\Lambda}_{YY} \end{bmatrix} \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}_X \\ \mathbf{y} - \boldsymbol{\mu}_Y \end{bmatrix} \\ = (\mathbf{x} - \boldsymbol{\mu}_X)^T \boldsymbol{\Lambda}_{XX} (\mathbf{x} - \boldsymbol{\mu}_X) + (\mathbf{x} - \boldsymbol{\mu}_X)^T \boldsymbol{\Lambda}_{XY} (\mathbf{y} - \boldsymbol{\mu}_Y) \\ + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Lambda}_{YX} (\mathbf{x} - \boldsymbol{\mu}_X) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Lambda}_{YY} (\mathbf{y} - \boldsymbol{\mu}_Y) $$

Substitute the expressions for $\boldsymbol{\Lambda}_{XX}$, $\boldsymbol{\Lambda}_{XY}$, and $\boldsymbol{\Lambda}_{YX}$ in terms of $\boldsymbol{\Sigma}$ blocks. After significant algebraic manipulation and completing the square with respect to $\mathbf{x}$, the exponent term can be rewritten as:

$$ -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) = -\frac{1}{2} [ (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) ] $$

Where:

*   $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
*   $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$

Now, let's substitute this back into the expression for $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$:

$$ f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} [ (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) ] \right) \\ \times \frac{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}{1} \exp\left( \frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right) $$

Notice that the term involving $(\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)$ cancels out. We are left with:

$$ f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right) $$

The determinant relationship $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\boldsymbol{\Sigma}_{X\mid Y}|$ is used to simplify the normalization constant.

Thus, the conditional distribution $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is indeed a multivariate Gaussian distribution with:

*   **Conditional Mean:** $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$
*   **Conditional Covariance:** $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$

This completes the derivation.

### Setup: Define the parameters of the random variables

### 1. Joint Distribution

**1.1 Show that $Y \sim \mathscr{N}(0, 5)$**

Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ will also be a normal random variable.

The mean of $Y$ is the sum of the means of $X$ and $Q$:
$E[Y] = E[X + Q] = E[X] + E[Q] = \mu_X + \mu_Q = 0 + 0 = 0$.

The variance of $Y$ is the sum of the variances of $X$ and $Q$ (due to independence):
$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = \sigma_X^2 + \sigma_Q^2 = 1 + 4 = 5$.

Therefore, $Y \sim \mathscr{N}(0, 5)$.

**1.2 Find the covariance $\text{Cov}(X, Y)$**

$\text{Cov}(X, Y) = \text{Cov}(X, X + Q)$
By linearity of covariance:
$\text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q)$

We know that $\text{Cov}(X, X) = \text{Var}(X) = \sigma_X^2 = 1$.
Since $X$ and $Q$ are independent, $\text{Cov}(X, Q) = 0$.

So, $\text{Cov}(X, Y) = 1 + 0 = 1$.

Let's verify these calculations with code.

### 2. Information Geometry

**Explain why $X$ is not $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?**

**Why $X$ is not $\sigma(Y)$-measurable:**

A random variable $X$ is $\sigma(Y)$-measurable if knowing the value of $Y$ allows us to determine the value of $X$ exactly. In other words, there must exist a function $g$ such that $X = g(Y)$.

In this problem, $Y = X + Q$. Since $Q$ is a random variable independent of $X$ and has its own variance ($	ext{Var}(Q)=4$), knowing $Y$ does not uniquely determine $X$. Specifically, $X = Y - Q$. Because $Q$ is random, $X$ remains random even if $Y$ is known. We can only talk about the *conditional distribution* of $X$ given $Y$, not a deterministic value.

**Practical explanation:**

If the power meter shows $Y=5$, the engineer cannot know for certain whether this is primarily due to a high temperature ($X$ is high) or high mechanical load ($Q$ is high). For example:

*   If $X=5$ (high temperature) and $Q=0$ (normal mechanical load), then $Y=5$.
*   If $X=1$ (normal temperature) and $Q=4$ (very hard material), then $Y=5$.
*   If $X=0$ (normal temperature) and $Q=5$ (extremely hard material), then $Y=5$.

Since $Q$ contributes to $Y$ and is independent of $X$, the observation of $Y=5$ contains a mix of information from both $X$ and $Q$. The `σ(Y)` algebra captures all information obtainable from $Y$. Because $Y$ alone cannot perfectly separate the contributions of $X$ and $Q$, $X$ cannot be precisely determined by $Y$, and thus $X$ is not $\sigma(Y)$-measurable. The engineer faces inherent uncertainty about the true cause of the high power reading.

### 3. Conditional Expectation (The Best Estimate)

**Compute the conditional expectation $E[X \mid Y]$**

For jointly normal random variables, the conditional expectation $E[X \mid Y = y]$ is given by:

$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$

We have:
*   $\mu_X = 0$
*   $\mu_Y = 0$
*   $\text{Cov}(X,Y) = 1$
*   $\text{Var}(Y) = 5$

Substitute these values:

$E[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$

So, $\widehat{X} = E[X \mid Y] = \frac{1}{5}Y$.

**Express this as a "signal-to-total-variance" ratio.**

The coefficient $\frac{\text{Cov}(X,Y)}{\text{Var}(Y)}$ can be interpreted. Since $\text{Cov}(X,Y) = \text{Var}(X)$ (because $Y=X+Q$ and $Q$ is independent of $X$), we have:

$\frac{\text{Cov}(X,Y)}{\text{Var}(Y)} = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)}$

This is often referred to as the "signal-to-total-variance" ratio (or sometimes, depending on context, the "reliability" or "filtering coefficient"). It represents the proportion of the total variance in $Y$ that is attributable to $X$. In this case, it's $\frac{1}{1+4} = \frac{1}{5}$.

So, $E[X \mid Y] = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} Y = \frac{1}{5}Y$.

```python
# Calculate E[X | Y]
conditional_expected_X_given_Y_coeff = cov_X_Y / sigma_Y_sq

print(f"E[X | Y] = {conditional_expected_X_given_Y_coeff} * Y")
print(f"This is a ratio of Var(X) / (Var(X) + Var(Q)) = {sigma_X_sq} / ({sigma_X_sq} + {sigma_Q_sq}) = {sigma_X_sq / (sigma_X_sq + sigma_Q_sq)}")
```

### 4. Verification of the Residual

Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.

**4.1 Prove that $E[Z] = 0$.**

We need to show $E[X - E[X \mid Y]] = 0$. By linearity of expectation:
$E[Z] = E[X] - E[E[X \mid Y]]$

A fundamental property of conditional expectation (Law of Total Expectation) is $E[E[X \mid Y]] = E[X]$.

Therefore,
$E[Z] = E[X] - E[X] = 0$.

**4.2 Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).**

We need to show $\text{Cov}(Z, Y) = 0$. Substitute $Z = X - E[X \mid Y]$:

$\text{Cov}(Z, Y) = \text{Cov}(X - E[X \mid Y], Y)$

Using linearity of covariance:
$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \text{Cov}(E[X \mid Y], Y)$

We found $E[X \mid Y] = \frac{1}{5}Y$. Substitute this:

$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \text{Cov}(\frac{1}{5}Y, Y)$

Recall $\text{Cov}(aW, V) = a \text{Cov}(W, V)$ and $\text{Cov}(V, V) = \text{Var}(V)$:

$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \frac{1}{5}\text{Var}(Y)$

Substitute the known values:
$\text{Cov}(X,Y) = 1$
$\text{Var}(Y) = 5$

$\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$.

This confirms that the estimation error $Z$ is uncorrelated with the observation $Y$, which is a key property of the conditional expectation as the optimal linear estimator (and optimal estimator in general for Gaussian case).

```python
# Verification of E[Z] = 0
# E[Z] = E[X] - E[E[X|Y]] = E[X] - E[X] = 0 by Law of Total Expectation.

# Verification of Cov(Z, Y) = 0
# Z = X - (1/5)Y
# Cov(Z, Y) = Cov(X - (1/5)Y, Y)
#           = Cov(X, Y) - Cov((1/5)Y, Y)
#           = Cov(X, Y) - (1/5)Var(Y)

calculated_cov_Z_Y = cov_X_Y - (conditional_expected_X_given_Y_coeff * sigma_Y_sq)

print(f"Calculated Cov(Z, Y) = {calculated_cov_Z_Y}")
print(f"Since Cov(Z, Y) = 0, the error Z is uncorrelated with the observation Y.")
```

### 5. Information Theory: Differential Entropy

For a normal distribution $\mathscr{N}(\mu, \sigma^2)$, the differential entropy is given by:
$h(X) = \frac{1}{2} \log_2(2\pi e \sigma^2)$

**5.1 Compute the differential entropy $h(X)$.**

$X \sim \mathscr{N}(0, 1)$, so $\sigma_X^2 = 1$.

$h(X) = \frac{1}{2} \log_2(2\pi e \cdot 1) = \frac{1}{2} \log_2(2\pi e)$

**5.2 Compute the conditional differential entropy $h(X \mid Y)$.**

The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.

Substitute the known values:
$\sigma_X^2 = 1$
$\text{Cov}(X,Y) = 1$
$\text{Var}(Y) = 5$

$\sigma_{X|Y}^2 = 1 - \frac{1^2}{5} = 1 - \frac{1}{5} = \frac{4}{5} = 0.8$.

So, $(X \mid Y=y) \sim \mathscr{N}(\mu^*, \sigma_{X|Y}^2)$ where $\mu^*$ is the conditional mean (which depends on $y$), but the conditional variance is constant.

Therefore, the conditional differential entropy $h(X \mid Y)$ is:

$h(X \mid Y) = \frac{1}{2} \log_2(2\pi e \sigma_{X|Y}^2) = \frac{1}{2} \log_2(2\pi e \cdot 0.8)$

Let's perform the calculations.

```python
# Helper function for differential entropy of a normal distribution
def differential_entropy_normal(variance):
    if variance <= 0:
        return float('-inf') # Entropy is undefined or negative infinity for non-positive variance
    return 0.5 * math.log2(2 * math.pi * math.e * variance)

# 5.1 Compute h(X)
h_X = differential_entropy_normal(sigma_X_sq)
print(f"h(X) = {h_X:.4f} bits")

# 5.2 Compute h(X | Y)
# Calculate conditional variance
sigma_X_given_Y_sq = sigma_X_sq - (cov_X_Y**2 / sigma_Y_sq)
h_X_given_Y = differential_entropy_normal(sigma_X_given_Y_sq)

print(f"Conditional variance σ²(X|Y) = {sigma_X_given_Y_sq:.4f}")
print(f"h(X | Y) = {h_X_given_Y:.4f} bits")
```

### 6. Mutual Information

**Calculate $I(X; Y) = h(X) - h(X \mid Y)$**

$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \log_2(2\pi e) - \frac{1}{2} \log_2(2\pi e \cdot 0.8)$
$I(X; Y) = \frac{1}{2} (\log_2(2\pi e) - \log_2(2\pi e \cdot 0.8))$
$I(X; Y) = \frac{1}{2} \log_2\left(\frac{2\pi e}{2\pi e \cdot 0.8}\right) = \frac{1}{2} \log_2\left(\frac{1}{0.8}\right) = \frac{1}{2} \log_2(1.25)$

**What percentage of the "uncertainty" about the temperature is removed by observing the power meter?**

This percentage is given by $\frac{I(X; Y)}{h(X)} \times 100\%$.

**How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?**

Let's recalculate the relevant quantities with the new $\sigma_Q^2 = 100$.

New $\sigma_Q^2 = 100$.
Original $\sigma_X^2 = 1$.

New $\text{Var}(Y) = \sigma_X^2 + \sigma_Q^2 = 1 + 100 = 101$.
New $\text{Cov}(X, Y) = \text{Var}(X) = 1$.

New conditional variance $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{101} = 1 - \frac{1}{101} = \frac{100}{101}$.

New $h(X \mid Y) = \frac{1}{2} \log_2(2\pi e \cdot \frac{100}{101})$.

New $I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \log_2(101/100)$.

Let's calculate these values.

```python
# 6.1 Calculate Mutual Information I(X; Y)
I_X_Y = h_X - h_X_given_Y
print(f"I(X; Y) = {I_X_Y:.4f} bits")

# 6.2 Percentage of uncertainty removed
percentage_removed = (I_X_Y / h_X) * 100
print(f"Percentage of uncertainty about X removed by observing Y = {percentage_removed:.2f}%")

print("\n--- Recalculating with Var(Q) = 100 ---")
# New parameters
sigma_Q_sq_new = 100
sigma_Y_sq_new = sigma_X_sq + sigma_Q_sq_new # 1 + 100 = 101
cov_X_Y_new = sigma_X_sq # Still 1

# New conditional variance
sigma_X_given_Y_sq_new = sigma_X_sq - (cov_X_Y_new**2 / sigma_Y_sq_new)

# New conditional differential entropy
h_X_given_Y_new = differential_entropy_normal(sigma_X_given_Y_sq_new)

# New Mutual Information
I_X_Y_new = h_X - h_X_given_Y_new

# New percentage removed
percentage_removed_new = (I_X_Y_new / h_X) * 100

print(f"New Var(Y) = {sigma_Y_sq_new}")
print(f"New Cov(X,Y) = {cov_X_Y_new}")
print(f"New conditional variance σ²(X|Y) = {sigma_X_given_Y_sq_new:.4f}")
print(f"New h(X | Y) = {h_X_given_Y_new:.4f} bits")
print(f"New I(X; Y) = {I_X_Y_new:.4f} bits")
print(f"New percentage of uncertainty about X removed by observing Y = {percentage_removed_new:.2f}%")

print("\nInterpretation of change:")
print("When the mechanical noise Q increases (Var(Q) goes from 4 to 100), the conditional variance σ²(X|Y) increases significantly. This means that after observing Y, there is more uncertainty remaining about X. Consequently, the mutual information I(X;Y) decreases, indicating that Y provides less information about X. The percentage of uncertainty removed drops from 10% to approximately 0.65%, meaning the power meter becomes much less effective at inferring the true internal temperature due to the overwhelming noise.")
```

### 7. Diagnostic Interpretation

**The factory manager says: "Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself." Using your result from Question 3, explain why $Y$ is a biased estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).**

**Why $Y$ is a biased estimator of $X$:**

An estimator $\widehat{X}$ for $X$ is unbiased if $E[\widehat{X}] = E[X]$.

If we were to use $Y$ as the estimator for $X$, i.e., $\widehat{X} = Y$, then:
$E[\widehat{X}] = E[Y] = E[X + Q] = E[X] + E[Q]$.

Since $E[X]=0$ and $E[Q]=0$, in this specific problem, $E[Y] = 0$, so $E[Y]=E[X]$. This implies that $Y$ is actually an *unbiased* estimator of $X$ in terms of its mean value. However, the manager's statement implies using the *value* of $Y$ directly as an estimate for $X$. While the *average* of $Y$ might match the *average* of $X$, the individual observations $Y$ are certainly not the best estimate for $X$ on a per-sample basis, as we'll see with the shrinking effect.

More generally, if $E[Q] \neq 0$, then $Y$ would be a biased estimator for $X$ (e.g., if $Q$ represented a constant offset).

**Why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero):**

From Question 3, we found $E[X \mid Y] = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} Y$.

Let $k = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)}$. In our case, $k = \frac{1}{1+4} = \frac{1}{5}$.

So, $E[X \mid Y] = kY = \frac{1}{5}Y$.

*   Since $X$ is normally distributed with mean 0, we can think of 0 as the "default" or "expected" temperature deviation in the absence of any other information.
*   The coefficient $k = \frac{1}{5}$ is between 0 and 1 (because $\text{Var}(Q) > 0$).
*   This means that for any observed value of $Y$, the estimated value of $X$ ($E[X \mid Y]$) will be $1/5$ of $Y$. If $Y$ is, say, $5$, the estimated $X$ is $1$. If $Y$ is $-10$, the estimated $X$ is $-2$. In both cases, the estimate for $X$ is closer to 0 than the observed $Y$ value. This is known as **shrinkage**.

**Why shrinkage occurs:**

The shrinkage effect arises because the observation $Y$ is noisy ($Y = X + Q$). When we observe a high (or low) value of $Y$, it could be due to a truly high (or low) $X$, or it could be partly due to noise $Q$. Since $Q$ has a mean of zero, and we have no reason to expect it to be systematically high or low, our best strategy is to assume that some portion of the extreme value in $Y$ is due to the noise $Q$ and thus 'regress' the estimate of $X$ back towards its own mean (which is 0). The magnitude of shrinkage is greater when the noise variance ($	ext{Var}(Q)$) is large relative to the signal variance ($	ext{Var}(X)$).

In our case, $	ext{Var}(X)=1$ and $	ext{Var}(Q)=4$. Since the noise variance is 4 times larger than the signal variance, a large portion of any deviation in $Y$ is attributed to $Q$, and therefore the estimate for $X$ is shrunk significantly towards 0.
