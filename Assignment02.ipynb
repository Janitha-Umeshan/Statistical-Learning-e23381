In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



1.Construct a probability space  (Ω,F,P)  suitable for this model.
Sample Space  Ω : The set of all possible outcomes for the number of emergency transformer failures. Based on the problem description,  X  takes values in the set  {0,1,2,3} . So, we can define  Ω={0,1,2,3} .

Event Space  F : Since  Ω  is a finite set, the event space  F  is the power set of  Ω , denoted by  P(Ω) . This includes all possible subsets of  Ω .

Probability Measure  P : The probabilities are given directly for each outcome in  Ω :

P({0})=0.50
P({1})=0.30

P({3})=0.05

2.Define  X  as a random variable on this space.

3.Determine the probability mass function of  X .
The probability mass function (PMF)  pX(x)  is given by the problem statement:

pX(0)=P(X=0)=0.50   
pX(1)=P(X=1)=0.30   
pX(2)=P(X=2)=0.15   
pX(3)=P(X=3)=0.05

pX(0)+pX(1)+pX(2)+pX(3)=1


```python
# Define the probability mass function (PMF)
pmf = {0: 0.50, 1: 0.30, 2: 0.15, 3: 0.05}

# 4. Compute the expectation E[X]
E_X = sum(x * p for x, p in pmf.items())
print(f"Expectation E[X]: {E_X:.2f}")

# 5. Compute the variance Var(X)
# E[X^2] = sum(x^2 * p)
E_X_squared = sum(x**2 * p for x, p in pmf.items())
Var_X = E_X_squared - (E_X)**2
print(f"Variance Var(X): {Var_X:.2f}")

# 6. Find P(X >= 1) and P(X >= 2)
P_X_ge_1 = sum(p for x, p in pmf.items() if x >= 1)
P_X_ge_2 = sum(p for x, p in pmf.items() if x >= 2)

print(f"P(X >= 1): {P_X_ge_1:.2f}")
print(f"P(X >= 2): {P_X_ge_2:.2f}")
```

Given a Triangular distribution with parameters:  a=4  (shortest time),  b=10  (longest time), and  c=6  (most likely time).

1.Specify the support of  X .
The support of the random variable  X  is the interval between the shortest and longest possible times. Therefore, the support of  X  is  [a,b]=[4,10]  hours.

2.Write down the probability density function  fX(x) .
The probability density function (PDF) for a Triangular distribution with parameters  a ,  b , and  c  is given by:

$$ f_X(x) = \begin{cases}
\frac{2(x-a)}{(b-a)(c-a)}, & \text{for } a \le x \le c \\
\frac{2(b-x)}{(b-a)(b-c)}, & \text{for } c < x \le b \\
0, & \text{otherwise}
\end{cases} $$


Substituting the given values  a=4 ,  b=10 , and  c=6 :

For  4≤x≤6 : $$ f_X(x) = \frac{2(x-4)}{(10-4)(6-4)} = \frac{2(x-4)}{6 \times 2} = \frac{2(x-4)}{12} = \frac{x-4}{6} $$

For  6<x≤10 : $$ f_X(x) = \frac{2(10-x)}{(10-4)(10-6)} = \frac{2(10-x)}{6 \times 4} = \frac{2(10-x)}{24} = \frac{10-x}{12} $$

So the PDF is:

$$ f_X(x) = \begin{cases}
\frac{x-4}{6}, & \text{for } 4 \le x \le 6 \\
\frac{10-x}{12}, & \text{for } 6 < x \le 10 \\
0, & \text{otherwise}
\end{cases} $$

3. Verify that  fX  integrates to  1  over its support.
We need to compute  ∫104fX(x)dx :

$$ \int_4^{10} f_X(x)\,dx = \int_4^6 \frac{x-4}{6}\,dx + \int_6^{10} \frac{10-x}{12}\,dx $$

**First integral:**
$$ \int_4^6 \frac{x-4}{6}\,dx = \frac{1}{6} \left[ \frac{x^2}{2} - 4x \right]_4^6 = \frac{1}{6} \left[ \left(\frac{36}{2} - 24\right) - \left(\frac{16}{2} - 16\right) \right] $$
$$ = \frac{1}{6} \left[ (18 - 24) - (8 - 16) \right] = \frac{1}{6} \left[ (-6) - (-8) \right] = \frac{1}{6} (2) = \frac{2}{6} = \frac{1}{3} $$

**Second integral:**
$$ \int_6^{10} \frac{10-x}{12}\,dx = \frac{1}{12} \left[ 10x - \frac{x^2}{2} \right]_6^{10} = \frac{1}{12} \left[ \left(100 - \frac{100}{2}\right) - \left(60 - \frac{36}{2}\right) \right] $$
$$ = \frac{1}{12} \left[ (100 - 50) - (60 - 18) \right] = \frac{1}{12} \left[ 50 - 42 \right] = \frac{1}{12} (8) = \frac{8}{12} = \frac{2}{3} $$

**Summing the integrals:**
$$ \int_4^{10} f_X(x)\,dx = \frac{1}{3} + \frac{2}{3} = 1 $$


The integral is indeed equal to  1 , so  fX(x)  is a valid probability density function.

```python
import math

# Given parameters for the Triangular distribution
a = 4
b = 10
c = 6

# 4. Compute the expectation E[X]
E_X = (a + b + c) / 3
print(f"Expectation E[X]: {E_X:.4f}")

# 5. Compute the variance Var(X)
Var_X = (a**2 + b**2 + c**2 - a*b - a*c - b*c) / 18
print(f"Variance Var(X): {Var_X:.4f}")

# Define the CDF for a Triangular distribution
def triangular_cdf(x, a, b, c):
    if x < a:
        return 0
    elif a <= x <= c:
        return (x - a)**2 / ((b - a) * (c - a))
    elif c < x <= b:
        return 1 - (b - x)**2 / ((b - a) * (b - c))
    else: # x > b
        return 1

# 8. Compute the probability that the installation takes at most 5 hours, P(X <= 5)
P_X_le_5 = triangular_cdf(5, a, b, c)
print(f"P(X <= 5): {P_X_le_5:.4f}")

# 9. Compute the probability that the installation takes between 5 and 8 hours, P(5 <= X <= 8)
P_X_le_8 = triangular_cdf(8, a, b, c)
P_5_le_X_le_8 = P_X_le_8 - P_X_le_5
print(f"P(5 <= X <= 8): {P_5_le_X_le_8:.4f}")
```

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

Solution for Q7: PERT Expectation Matching Most Likely Time

Given the parameters for the PERT distribution:

Minimum time:  a=5  days
Most likely time:  c=11  days
And the condition that the expected time is the same as the most likely time:  E[X]=c=11  days.

```python
import math

# Given parameters
a = 5
c = 11
E_X = c  # Expected time is the same as the most likely time

# 1. Determine the value of the maximum time b.
# We know the formula for E[X]: E[X] = (a + 4*c + b) / 6
# Rearrange to solve for b:
# 6 * E[X] = a + 4*c + b
# b = 6 * E[X] - a - 4*c
b = 6 * E_X - a - 4 * c

print(f"Calculated maximum time b: {b:.0f} days")

# 2. What happens to the variance?
# Variance approximation formula: Var(X) = (E[X] - a)(b - E[X]) / 7
# Substitute the calculated b and given values
Var_X = ((E_X - a) * (b - E_X)) / 7

print(f"Calculated Variance Var(X): {Var_X:.4f}")
```

1.Determine the value of the maximum time  b .
The formula for the expectation of a PERT distribution is:
$$ E[X] = \frac{a + 4c + b}{6} $$

We are given $a=5$, $c=11$, and $E[X]=c=11$. Substitute these values into the formula:
$$ 11 = \frac{5 + 4(11) + b}{6} $$
$$ 11 = \frac{5 + 44 + b}{6} $$
$$ 11 = \frac{49 + b}{6} $$
Multiply both sides by 6:
$$ 66 = 49 + b $$
Solve for $b$:
$$ b = 66 - 49 $$
$$ b = 17 $$

So, the maximum time  b  is 17 days.

2.What happens to the variance?

The approximate formula for the variance of a PERT distribution is:

$$ \mathrm{Var}(X) = \frac{(E[X]-a)(b-E[X])}{7} $$

Substitute the values $a=5$, $b=17$, and $E[X]=11$ into the formula:
$$ \mathrm{Var}(X) = \frac{(11-5)(17-11)}{7} $$
$$ \mathrm{Var}(X) = \frac{(6)(6)}{7} $$
$$ \mathrm{Var}(X) = \frac{36}{7} \approx 5.1429 $$

Analysis of the variance: When the expected time  E[X]  is equal to the most likely time  c , and  b  is adjusted accordingly (in this case,  b=17  for  a=5,c=11,E[X]=11 ), the variance remains a positive value. It does not become zero. This indicates that even if the average completion time coincides with the most likely time, there is still uncertainty and variability in the task duration. The spread of possible outcomes between the minimum ( a=5 ) and maximum ( b=17 ) values ensures a non-zero variance. In this specific case, the variance is approximately  5.14 , which implies a standard deviation of  5.14−−−−√≈2.27  days, suggesting a moderate spread around the 11-day expectation.
